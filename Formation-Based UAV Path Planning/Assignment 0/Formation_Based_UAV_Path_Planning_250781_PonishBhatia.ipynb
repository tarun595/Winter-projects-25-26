## Your Solution 1

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 2500)
r = 1 + 3*np.sin(6*theta)

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
def r(theta):
    return 1 + 3*np.sin(6*theta)
def rprime(theta):
    return 18*np.cos(6*theta)
def integrate(f, a, b, n):
    h = (b - a) / n
    x = a + h * np.arange(n + 1)
    fx = f(x)
    return np.sum(fx)*h

theta0 = 0
theta1 = 2*np.pi
n = 10000000

area_integrand = lambda t: 0.5 * r(t)**2
arc_integrand = lambda t: np.sqrt(r(t)**2 + rprime(t)**2)

area = integrate(area_integrand, theta0, theta1, n)
arclength = integrate(arc_integrand, theta0, theta1, n)

print("Area =", round(area,3))
print("Arc length =", round(arclength,3))
```

## Question 3
The energy required to get from point $\vec{r}_1$ to point $\vec{r}_2$ for a drone is given by

$$ E = \alpha \int_{C} \left| \frac{d\vec{r}}{dt} \right| dt - \int_C \vec{F} \cdot \frac{d\vec{r}}{dt}dt $$



3.1) Suppose that $\alpha=5$ and our start and ends points are $\vec{r}_1 = (0,0)$ and $\vec{r}_2 = (0, 10)$. On this particular day, the wind produces a force field $\vec{F} = (0, -2/(x+1)^2)$. Approximate the optimal value of $A$ (A>0) in $\vec{r}(t) = A\sin(\pi t/10)\hat{x} + t\hat{y}$ that minimizes the **work** using only numpy arrays .

3.2) Plot Energy as a function of A . (E(A) v/s A ) ? (Comment about its shape)

## Your Solution 3

```python
alpha = 5

t = np.linspace(0, 10, 50000)
dt = t[1] - t[0]

def energy(A):
    x = A * np.sin(np.pi * t / 10)
    dxdt = A * (np.pi / 10) * np.cos(np.pi * t / 10)

    speed = np.sqrt(dxdt**2 + 1)
    force_y = -2 / (x + 1)**2

    E1 = alpha * np.sum(speed) * dt
    E2 = np.sum(force_y) * dt

    return E1 - E2

A_vals = np.linspace(0.01, 5, 500)
E_vals = np.array([energy(A) for A in A_vals])

A_opt = A_vals[np.argmin(E_vals)]
print(A_opt)
```

```python
plt.figure()
plt.plot(A_vals, E_vals)
plt.xlabel("A")
plt.ylabel("Energy E(A)")
plt.title("Energy vs Amplitude A")
plt.show()

#the graph makes it look like the energy decreases first to achieve a minimum, and then starts increasing exponentially to infinity
#the high energy near zero amplitude is required to move against the wind, the high energy for high amplitudes is required to cover the large distance due to high amplitude
#near A=1.43, there seems to be a resonance where both effects minimise each other the best they can, and we get the energy consumption minimum at that amplitude (1.43)
```

## Question 4

You decide to install a series of camera drones along a straight path to spy on a suspicious person located at marker 31

You buy four drones and they are evenly spaced starting from mile marker 16 and ending on mile marker 25.

You buy another four drones starting on mile marker 29 and ending on mile marker 34. (also evenly spaced)

A drone can face only one direction at a time.
In order, from marker 16 to 34, each drone's orientation : N,S,E,W,S,E,W,N. (where N,S,E,W are cardinal directions)

Determine how far each drone facing E is from the person? (Solve using Numpy only)

## Your Solution 4

```python
drones1 = np.linspace(16, 25, 4)
drones2 = np.linspace(29, 34, 4)
```

```python
drones = np.concatenate([drones1, drones2])
```

```python
directions = np.array(['N', 'S', 'E', 'W', 'S', 'E', 'W', 'N'])
```

```python
east = directions == 'E'
```

```python
person = 31
east_facing = drones[east]
distances = np.abs(east_facing - person)
```
