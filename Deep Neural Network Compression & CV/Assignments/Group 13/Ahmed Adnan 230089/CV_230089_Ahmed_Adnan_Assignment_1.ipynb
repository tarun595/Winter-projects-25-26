````python
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import os


plt.rcParams['figure.figsize'] = (10,6)


# -------------------------------
# Helper functions
# -------------------------------


def imshow_rgb(img, title=None):
# img is assumed to be in RGB order (H,W,3)
    plt.figure()
    plt.imshow(img.astype(np.uint8))
    if title: plt.title(title)
    plt.axis('off')

def imshow_gray(img, title=None):
   plt.figure()
   plt.imshow(img, cmap='gray')
   if title: plt.title(title)
   plt.axis('off')

def show_side_by_side(images, titles=None, cmap=None):
   n = len(images)
   fig, axs = plt.subplots(1, n, figsize=(4*n,4))
   if n==1:
     axs = [axs]
   for i,ax in enumerate(axs):
      if cmap:
       ax.imshow(images[i], cmap=cmap)
      else:
          ax.imshow(images[i].astype(np.uint8))
      if titles: ax.set_title(titles[i])
      ax.axis('off')
   plt.show()

def to_uint8(img):
# scale and convert to uint8 safely
   img = np.real(img)
   img = np.clip(img, 0, 255)
   return img.astype(np.uint8)
````

````python
# Default path provided by developer
img_path = '/content/CHICAGO-PANORAMIC-SKYLINE-NIGHT-10.19-0107-1200PX.jpg'
if not os.path.exists(img_path):
  raise FileNotFoundError(f"Default image not found at {img_path}. Replace img_path with your image or embed base64.")


# Read with OpenCV (BGR) then convert to RGB for display
bgr = cv2.imread(img_path, cv2.IMREAD_COLOR)
if bgr is None:
  raise ValueError("cv2 failed to read the image. Make sure the path is correct and file is a valid image.")


rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)


# Display the RGB image
imshow_rgb(rgb, 'Original RGB Image (Part A)')


# Convert to grayscale using the weighted sum formula (NumPy) as requested.
# Common formula: Y = 0.299 R + 0.587 G + 0.114 B
weights = np.array([0.299, 0.587, 0.114])
# multiply channels
gray_np = np.dot(rgb[..., :3], weights)


# Convert using cv2.cvtColor for comparison
gray_cv2 = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY) # OpenCV reads BGR so this is equivalent


# Display both grays
show_side_by_side([gray_np, gray_cv2], titles=['Grayscale (NumPy weighted sum)', 'Grayscale (cv2.cvtColor)'], cmap='gray')


# Print array shapes
print('RGB shape:', rgb.shape)
print('Grayscale (NumPy) shape:', gray_np.shape)
print('Grayscale (cv2) shape:', gray_cv2.shape)


# Print a few pixel values (top-left corner and a center pixel)
h,w = gray_np.shape
coords = [(0,0), (h//2, w//2), (10,10)]
print('\nSome pixel values (R,G,B) and grayscale values:')
for (r,c) in coords:
   print(f'Pixel ({r},{c}) - RGB: {rgb[r,c,:]} | gray_np: {gray_np[r,c]:.2f} | gray_cv2: {gray_cv2[r,c]:.2f}')


# Note: Number of channels is different. RGB has 3 channels (shape HxWx3), grayscale has single channel (HxW).
````

````python
img_float = gray_np.astype(np.float32)


# B1: compute fft2 and shift
f = np.fft.fft2(img_float)
fshift = np.fft.fftshift(f)


# B2: magnitude, phase and magnitude spectrum
magnitude = np.abs(fshift)
phase = np.angle(fshift)


# For display, magnitude spectrum is typically log-scaled so small values are visible
magnitude_spectrum = 20 * np.log(np.clip(magnitude, a_min=1e-8, a_max=None))


# show
show_side_by_side([magnitude_spectrum, phase], titles=['Magnitude Spectrum (log scaled)', 'Phase (radians)'], cmap='gray')


# Print some stats
print('FFT arrays shapes: fshift:', fshift.shape, 'magnitude:', magnitude.shape, 'phase:', phase.shape)
````

````python
rows, cols = img_float.shape
crow, ccol = rows//2 , cols//2


# Function to create circular low-pass filter mask
def circular_mask(shape, radius):
   rows, cols = shape
   crow, ccol = rows//2, cols//2
   Y, X = np.ogrid[:rows, :cols]
   dist_from_center = np.sqrt((X - ccol)**2 + (Y - crow)**2)
   mask = dist_from_center <= radius
   return mask.astype(np.float32)


# Radii for LPF
radii = [10, 30, 80]
reconstructed_lpf = []


for r in radii:
   mask = circular_mask(img_float.shape, r)
# apply mask to fshift (element-wise)
   fshift_masked = fshift * mask
# inverse
   f_ishift = np.fft.ifftshift(fshift_masked)
   img_back = np.fft.ifft2(f_ishift)
   img_back = np.abs(img_back)
   reconstructed_lpf.append(img_back)


# Show LPF results (smaller radius -> stronger blur because more high frequencies are removed)
titles = [f'LPF radius={r}' for r in radii]
show_side_by_side([to_uint8(im) for im in reconstructed_lpf], titles=titles, cmap='gray')


# Explanation in comment:
# When the LPF radius is small, only very low-frequency components (coarse structure) are kept.
# High-frequency components (edges, fine details, texture) are removed leading to stronger blurring.
# As radius increases, more high-frequency content is retained and image becomes sharper.


# C2: HPF
# High-pass filter can be created as 1 - LPF
r_hpf = 30
lpf_mask = circular_mask(img_float.shape, r_hpf)
hpf_mask = 1.0 - lpf_mask
fshift_hpf = fshift * hpf_mask
f_ishift = np.fft.ifftshift(fshift_hpf)
img_hpf = np.fft.ifft2(f_ishift)
img_hpf = np.abs(img_hpf)


imshow_gray(to_uint8(img_hpf), title=f'High-pass filtered (radius={r_hpf})')


# C3: Square LPF mask centered with side length 50
side = 50
mask_square = np.zeros(img_float.shape, dtype=np.float32)
start_r = rows//2 - side//2
start_c = cols//2 - side//2
mask_square[start_r:start_r+side, start_c:start_c+side] = 1.0


fshift_square = fshift * mask_square
f_ishift = np.fft.ifftshift(fshift_square)
img_square = np.fft.ifft2(f_ishift)
img_square = np.abs(img_square)


imshow_gray(to_uint8(img_square), title=f'Square LPF side={side}')


````

````python
# D1:
eps = 1e-9

def normalize_to_uint8(arr, eps=1e-9):
    """
    Normalize a real-valued array to [0,255] and convert to uint8.
    Uses np.ptp(arr) (NumPy 2.0+ compatible) for peak-to-peak.
    """
    arr = np.real(arr)
    a_min = arr.min()
    a_ptp = np.ptp(arr)  # replaced arr.ptp() with np.ptp for NumPy 2.0+
    if a_ptp < eps:
        # nearly-constant image  just clip and convert
        return to_uint8(arr)
    normalized = 255 * (arr - a_min) / (a_ptp + eps)
    return to_uint8(normalized)

# D1: Phase-only reconstruction
# Remove amplitude, keep only phase (unit magnitude complex numbers).
phase_only = fshift / (np.abs(fshift) + eps)   # equivalent to np.exp(1j * np.angle(fshift)) except where magnitude==0

# Inverse transform to spatial domain
f_ishift = np.fft.ifftshift(phase_only)
img_phase_only = np.fft.ifft2(f_ishift)
img_phase_only = np.real(img_phase_only)

# Display (normalized to 0-255)
imshow_gray(normalize_to_uint8(img_phase_only), title='Phase-only Reconstruction (normalized)')

# Note:
# - np.angle(fshift) returns the phase angles. To build a complex array with those phases you can do:
#     phase_exp = np.exp(1j * np.angle(fshift))
#   which is numerically equivalent to phase_only above (except for tiny differences where magnitude==0).
# - phase_only = fshift / abs(fshift) keeps only the phase information (unit magnitude), discarding amplitude.

# Try reconstruction using np.exp(1j * np.angle(fshift)) to show equivalence (optional)
phase_exp = np.exp(1j * np.angle(fshift))
f_ishift = np.fft.ifftshift(phase_exp)
img_phase_from_angle = np.fft.ifft2(f_ishift)
img_phase_from_angle = np.real(img_phase_from_angle)
# They should appear essentially the same
#imshow_gray(normalize_to_uint8(img_phase_from_angle), title='Phase from np.angle Reconstruction (normalized)')

# D2: Magnitude-only reconstruction
# Use magnitude but set phase to zero (i.e. assume all frequency bins have 0 phase)
mag_only = np.abs(fshift)
mag_only_complex = mag_only * np.exp(1j * 0)   # same as mag_only + 0j

# Inverse transform
f_ishift = np.fft.ifftshift(mag_only_complex)
img_mag_only = np.fft.ifft2(f_ishift)
img_mag_only = np.real(img_mag_only)

# Display (normalized to 0-255)
imshow_gray(normalize_to_uint8(img_mag_only), title='Magnitude-only Reconstruction (zero phase, normalized)')

# Role of logarithm:
# The FFT magnitude spans many orders of magnitude; log-compression (e.g., 20*log10(magnitude)) reduces dynamic range
# so small spectral features become visible in the plotted magnitude spectrum. Without log-scaling, the display would
# be dominated by a few very large coefficients (often low frequencies) and most detail would be invisible.
````
