#### Task 3 - Gaussian Blur

1. Implement a Gaussian kernel generator:
The function signature should be <code>gaussian_kernel(ksize, sigma)</code>
It should return a numpy matrix that is the normalsied gaussian kernel with dimensions ksize x ksize and the 2D gaussian distribution calculation done with standard deviation as sigma. ````python

````python
def show(img_array, title="", cmap=None):
    plt.figure(figsize=(5, 5))
    plt.imshow(img_array, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()


img_gray = np.dot(img_part5[...,:3], [0.2989, 0.5870, 0.1140])
img_gray = img_gray.astype(np.uint8)


def convolve(image, kernel):

    kernel = np.flipud(np.fliplr(kernel))


    k_h, k_w = kernel.shape


    pad_h = k_h // 2
    pad_w = k_w // 2


    if len(image.shape) == 2:

        padded_img = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='reflect')
        output = np.zeros_like(image, dtype=float)


        for i in range(image.shape[0]):
            for j in range(image.shape[1]):

                roi = padded_img[i:i+k_h, j:j+k_w]

                output[i, j] = np.sum(roi * kernel)

    else:

        padded_img = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w), (0, 0)), mode='reflect')
        output = np.zeros_like(image, dtype=float) # Use float for output

        for c in range(image.shape[2]):
            for i in range(image.shape[0]):
                for j in range(image.shape[1]):
                    roi = padded_img[i:i+k_h, j:j+k_w, c]
                    output[i, j, c] = np.sum(roi * kernel)

    return output


def average_blur(image, ksize):

    kernel = np.ones((ksize, ksize)) / (ksize * ksize)
    return convolve(image, kernel)


avg_blurred_img = average_blur(img_part5, 5) # 5x5 blur
show(avg_blurred_img.astype(np.uint8), "Average Blur (5x5)")



def gaussian_kernel(ksize, sigma):

    ax = np.linspace(-(ksize - 1) / 2., (ksize - 1) / 2., ksize)
    xx, yy = np.meshgrid(ax, ax)


    kernel = np.exp(-0.5 * (np.square(xx) + np.square(yy)) / np.square(sigma))


    return kernel / np.sum(kernel)


g_kernel = gaussian_kernel(5, 1.0)
gauss_blurred_img = convolve(img_part5, g_kernel)

show(gauss_blurred_img.astype(np.uint8), "Gaussian Blur (sigma=1)")
````

#### Task - Frequency Analysis (LPF & HPF Verification)
Compute the magnitude spectrum of the original image, gaussian blurred image and the unsharp masking image. Now, play with the alpha value in sharpening and gaussian kernel and notice how the magntude spectrum changes.

````python


def get_magnitude_spectrum(img):

    f = np.fft.fft2(img)

    fshift = np.fft.fftshift(f)

    magnitude_spectrum = 20 * np.log(np.abs(fshift) + 1)
    return magnitude_spectrum


mag_original = get_magnitude_spectrum(img_gray)


g_k = gaussian_kernel(9, 2.0)
img_blurred_severe = convolve(img_gray, g_k)
mag_blur = get_magnitude_spectrum(img_blurred_severe)


mag_sharp = get_magnitude_spectrum(sharpened_usm)

# Plotting
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.imshow(mag_original, cmap='gray')
plt.title("Original Spectrum")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(mag_blur, cmap='gray')
plt.title("Gaussian Blur Spectrum")
plt.axis('off')


plt.subplot(1, 3, 3)
plt.imshow(mag_sharp, cmap='gray')
plt.title("Sharpened Spectrum")
plt.axis('off')


plt.show()
````
