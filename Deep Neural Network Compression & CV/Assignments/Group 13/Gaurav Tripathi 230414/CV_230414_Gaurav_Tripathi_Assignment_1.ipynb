B1

```python
# B1  Compute FFT and shifted FFT

import numpy as np
gray = gray_manual      # from Part A

# Ensure grayscale array is float
gray_f = gray.astype(np.float32)

# Step 1: 2D FFT
F = np.fft.fft2(gray_f)

# Step 2: Shift zero frequency to center
fshift = np.fft.fftshift(F)

print("FFT shape:", F.shape)
print("Shifted FFT shape:", fshift.shape)
```

B2

```python
# B2  Compute magnitude, phase, and display magnitude spectrum

import matplotlib.pyplot as plt

# Magnitude = absolute value of complex frequency
magnitude = np.abs(fshift)

# Phase = angle of complex frequency
phase = np.angle(fshift)

# Use log scale for better visualization
magnitude_spectrum = np.log1p(magnitude)

# Display
plt.figure(figsize=(6,5))
plt.imshow(magnitude_spectrum, cmap='inferno')
plt.title("Magnitude Spectrum (log scale)")
plt.axis("off")
plt.show()

# Print a few values
print("Magnitude sample values:")
print(magnitude[100:105, 100:105])

print("\nPhase sample values:")
print(phase[100:105, 100:105])
```

```python
# ===== C1: Circular LPFs with radii 10, 30, 80 =====
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Preconditions: gray (H,W) and fshift (H,W complex) exist. If not, create them:
# gray = np.array(img.convert('L'))          # fallback if needed
# fshift = np.fft.fftshift(np.fft.fft2(gray.astype(np.float32)))

H, W = gray.shape
cy, cx = H // 2, W // 2

# Choose radii (absolute pixel values). If image is small, scale them:
radii = [10, 30, 80]
# If you prefer image-size relative radii, you can use:
# radii = [int(min(H,W)*0.02), int(min(H,W)*0.06), int(min(H,W)*0.16)]

def circular_mask(H, W, center, radius):
    y = np.arange(H)[:, None]
    x = np.arange(W)[None, :]
    dist2 = (y - center[0])**2 + (x - center[1])**2
    return (dist2 <= radius**2).astype(np.float32)

reconstructions = []
masks = []
for r in radii:
    mask = circular_mask(H, W, (cy, cx), r)
    masks.append(mask)
    # Apply mask to centered FFT
    fshift_masked = fshift * mask
    # Reconstruct
    f_ishift = np.fft.ifftshift(fshift_masked)
    img_back = np.fft.ifft2(f_ishift)
    img_back = np.real(img_back)
    # Clip & convert for display
    img_back_clipped = np.clip(img_back, 0, 255).astype(np.uint8)
    reconstructions.append(img_back_clipped)

# Plot original grayscale + three reconstructions
plt.figure(figsize=(12, 3.5))
plt.subplot(1, 4, 1); plt.imshow(gray, cmap='gray'); plt.title('Original'); plt.axis('off')
for i, r in enumerate(radii):
    plt.subplot(1, 4, i+2)
    plt.imshow(reconstructions[i], cmap='gray')
    plt.title(f'LPF radius={r}')
    plt.axis('off')
plt.suptitle('C1: Circular LPFs  smaller radius => stronger blur', fontsize=14)
plt.show()

# Print short diagnostics
for r, mask in zip(radii, masks):
    kept = mask.sum()
    print(f"Radius {r} -> mask ones count = {int(kept)} ({kept/(H*W):.3%} of freq bins kept)")
```

c2

```python
# ===== C2: High-Pass Filter (HPF) =====
import numpy as np
import matplotlib.pyplot as plt

# We'll build an HPF by subtracting an LPF mask from 1
# Choose radius for low-frequency removal center (e.g., 30). You can change it.
hpf_radius = 30

lp_mask = circular_mask(H, W, (cy, cx), hpf_radius)
hpf_mask = 1.0 - lp_mask   # High-pass: zeros at center, ones at high freq

# Apply HPF to fshift
fshift_hpf = fshift * hpf_mask
# Reconstruct
f_ishift_hpf = np.fft.ifftshift(fshift_hpf)
img_hpf = np.fft.ifft2(f_ishift_hpf)
img_hpf = np.real(img_hpf)
img_hpf_clipped = np.clip(img_hpf, 0, 255).astype(np.uint8)

# Show original vs HPF result
plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(gray, cmap='gray'); plt.title('Original'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(img_hpf_clipped, cmap='gray'); plt.title(f'HPF (radius={hpf_radius})'); plt.axis('off')
plt.suptitle('C2: High-Pass Filter result (edges and high-frequency detail)', fontsize=12)
plt.show()

# Print diagnostics
print("HPF mask ones count:", int(hpf_mask.sum()), f"({hpf_mask.sum()/(H*W):.3%} of bins kept)")
```

c3

```python
# ===== C3: Square LPF mask (side = 50) =====
side = 50  # side length of the square low-pass window
# Ensure side is odd/even doesn't matter; compute half
half = side // 2

# Build square mask
square_mask = np.zeros((H, W), dtype=np.float32)
y0, y1 = max(0, cy-half), min(H, cy+half+ (1 if side%2 else 0))
x0, x1 = max(0, cx-half), min(W, cx+half+ (1 if side%2 else 0))
square_mask[y0:y1, x0:x1] = 1.0

# Apply
fshift_square = fshift * square_mask
f_ishift_square = np.fft.ifftshift(fshift_square)
img_square = np.fft.ifft2(f_ishift_square)
img_square = np.real(img_square)
img_square_clipped = np.clip(img_square, 0, 255).astype(np.uint8)

# Show results
plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.imshow(gray, cmap='gray'); plt.title('Original'); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(img_square_clipped, cmap='gray'); plt.title(f'Square LPF side={side}'); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(square_mask, cmap='gray'); plt.title('Square mask (frequency)'); plt.axis('off')
plt.suptitle('C3: Square LPF (central square kept)', fontsize=12)
plt.show()

# Diagnostics
print("Square mask ones:", int(square_mask.sum()), f"({square_mask.sum()/(H*W):.3%} kept)")
print("Square region coords y0,y1,x0,x1:", (y0,y1,x0,x1))
```

D1

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# ---------- D1: Phase-only reconstruction ----------

# Safety: avoid divide-by-zero (where magnitude is 0)
mag = np.abs(fshift)
eps = 1e-8
mag_safe = mag + eps

# Phase-only spectrum: complex numbers with magnitude = 1, phase same as original
phase_only = fshift / mag_safe

# NOTE (very important):
# phase_only contains COMPLEX VALUES on the unit circle: e^(j*theta)
# This is different from np.angle(fshift), which returns ONLY the real angle theta
# (a float array, not complex). For reconstruction, we need complex exponentials,
# not just the angle values themselves.

# Reconstruct from phase-only spectrum
f_ishift_phase = np.fft.ifftshift(phase_only)
img_phase = np.fft.ifft2(f_ishift_phase)
img_phase = np.real(img_phase)

# Normalize to [0, 255] for display (phase-only changes overall scaling/contrast)
img_phase_norm = img_phase - img_phase.min()
if img_phase_norm.max() > 0:
    img_phase_norm = img_phase_norm / img_phase_norm.max()
img_phase_255 = (img_phase_norm * 255).astype(np.uint8)

plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_phase_255, cmap='gray')
plt.title("Phase-only Reconstruction")
plt.axis("off")

plt.suptitle("D1: Reconstructing from PHASE only", fontsize=13)
plt.show()

# ---------- Extra: wrong reconstruction using np.angle(fshift) directly ----------

angle_only = np.angle(fshift)
# If we try to use angle_only directly as the spectrum:
f_ishift_angle = np.fft.ifftshift(angle_only)
img_angle_wrong = np.fft.ifft2(f_ishift_angle)
img_angle_wrong = np.real(img_angle_wrong)

img_angle_norm = img_angle_wrong - img_angle_wrong.min()
if img_angle_norm.max() > 0:
    img_angle_norm = img_angle_norm / img_angle_norm.max()
img_angle_255 = (img_angle_norm * 255).astype(np.uint8)

plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.imshow(img_phase_255, cmap='gray')
plt.title("Phase-only (correct method)")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_angle_255, cmap='gray')
plt.title("Using np.angle(fshift) as spectrum (incorrect)")
plt.axis("off")

plt.suptitle("D1: Comparing phase-only vs raw angle", fontsize=13)
plt.show()

# Code comments / explanation (to mention in your report):
# phase_only = fshift / |fshift| keeps ONLY the phase but as COMPLEX unit-magnitude
# coefficients, i.e. each frequency becomes e^(j*theta). This is a valid spectrum.
# np.angle(fshift) returns only theta (real numbers), so if we feed that directly
# to the inverse FFT, we are treating those angles as if they were magnitudes.
# That destroys the proper complex structure of the spectrum, so the reconstruction
# looks like a weird, low-contrast / noisy pattern rather than a recognizable image.
```

D2

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- D2: Magnitude-only reconstruction ----------

# Magnitude of spectrum
mag_only = np.abs(fshift)

# For reconstruction, we must combine magnitude with some phase.
# "Magnitude-only" usually means we set phase = 0 everywhere:
# spectrum_recon = |F(u,v)| * e^(j*0) = |F(u,v)|
mag_spec_zero_phase = mag_only * np.exp(1j * 0)

f_ishift_mag = np.fft.ifftshift(mag_spec_zero_phase)
img_mag = np.fft.ifft2(f_ishift_mag)
img_mag = np.real(img_mag)

# Normalize to [0,255] for viewing
img_mag_norm = img_mag - img_mag.min()
if img_mag_norm.max() > 0:
    img_mag_norm = img_mag_norm / img_mag_norm.max()
img_mag_255 = (img_mag_norm * 255).astype(np.uint8)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(img_mag_255, cmap='gray')
plt.title("Magnitude-only Reconstruction\n(phase = 0)")
plt.axis("off")

# For comparison: magnitude spectrum visualization with log scaling
magnitude = mag_only
magnitude_log = np.log1p(magnitude)   # log(1 + |F|) for visualization
plt.subplot(1,3,3)
plt.imshow(magnitude_log, cmap='inferno')
plt.title("Log Magnitude Spectrum\n(Visualization only)")
plt.axis("off")

plt.suptitle("D2: Magnitude-only reconstruction vs Magnitude Spectrum", fontsize=13)
plt.show()

# Explanation to put in comments:
# 1) During reconstruction we use mag_only as |F| and set phase=0. This removes
#    all structural phase information, so the reconstructed image usually looks
#    like a kind of symmetric blob / pattern, not the original scene.
#
# 2) The Magnitude Spectrum we often display uses log scaling:
#       magnitude_log = log(1 + |F|)
#    The raw magnitude |F| has a huge dynamic range: a few very large low-frequency
#    coefficients (near DC) and many tiny high-frequency ones. Without log, the
#    spectrum image would be almost all black except a bright spot at the center.
#    Logarithm compresses large values and stretches small ones, so we can see
#    mid- and high-frequency content.
#
# 3) The reconstructed magnitude-only image looks different from the displayed
#    Magnitude Spectrum because:
#    - The reconstruction is inverse FFT of a spectrum (|F| with zero phase) back
#      into the spatial domain, giving an image-like pattern.
#    - The Magnitude Spectrum plot is just a visualization of |F| in the frequency
#      domain, with log scaling, NOT a spatial image of the scene.
```
