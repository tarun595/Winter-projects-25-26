C1  Create 3 LPF (Low-Pass Filter) Masks


````python
import numpy as np
import matplotlib.pyplot as plt
rows, cols = gray_np.shape
crow, ccol = rows//2, cols//2
````

````python
def create_lpf_mask(radius):
    mask = np.zeros((rows, cols), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            if np.sqrt((i - crow)**2 + (j - ccol)**2) <= radius:
                mask[i, j] = 1
    return mask
````

````python
radii = [10, 30, 80]
for R in radii:
    mask = create_lpf_mask(R)
    f_filtered = fshift * mask
    f_ishift = np.fft.ifftshift(f_filtered)
    img_back = np.fft.ifft2(f_ishift)
    img_back = np.abs(img_back)
    plt.figure(figsize=(5,5))
    plt.imshow(img_back, cmap='gray')
    plt.title(f"LPF Output (radius = {R})")
    plt.axis("off")
    plt.show()
````

C2  Create High-Pass Filter (HPF) Mask

````python
def create_hpf_mask(radius):
    lpf = create_lpf_mask(radius)
    hpf = 1 - lpf
    return hpf
````

````python
R = 25

hpf_mask = create_hpf_mask(R)
f_filtered = fshift * hpf_mask

f_ishift = np.fft.ifftshift(f_filtered)
img_back_hpf = np.abs(np.fft.ifft2(f_ishift))

plt.figure(figsize=(5,5))
plt.imshow(img_back_hpf, cmap='gray')
plt.title("High-Pass Filter Output")
plt.axis("off")
plt.show()
````

C3  Square LPF Mask (Side = 50)

````python
square_size = 50
half = square_size // 2
mask_sq = np.zeros((rows, cols), dtype=np.uint8)
mask_sq[crow-half:crow+half, ccol-half:ccol+half] = 1
````

````python
f_filtered = fshift * mask_sq
f_ishift = np.fft.ifftshift(f_filtered)
img_back_sq = np.abs(np.fft.ifft2(f_ishift))
plt.figure(figsize=(5,5))
plt.imshow(img_back_sq, cmap='gray')
plt.title("Square LPF Mask Output")
plt.axis("off")
plt.show()
````

D1. Phase-only reconstruction

````python
phase_only = fshift / np.abs(fshift)
# Phase-only reconstruction
eps = 1e-8
phase_only = fshift / (np.abs(fshift) + eps)
# Reconstruction
phase_ishift = np.fft.ifftshift(phase_only)
phase_img = np.abs(np.fft.ifft2(phase_ishift))

plt.figure(figsize=(6,6))
plt.imshow(phase_img, cmap='gray')
plt.title("Phase-Only Reconstruction")
plt.axis("off")
plt.show()



````

````python
phase_only = fshift / np.abs(fshift)
np.angle(fshift)
````

np.angle(fshift) extracts only the phase values (in radians),
but does NOT produce a valid complex spectrum for reconstruction.

phase_only = fshift / |fshift|
creates a complex number whose magnitude = 1 and phase = original phase,
so it keeps the correct Fourier structure needed to reconstruct an image.

That is why reconstruction works with (fshift / |fshift|) but NOT with np.angle().


D2. Magnitude-Only Reconstruction

````python
mag_only = np.abs(fshift)
#  Magnitude-only reconstruction
mag_only = np.abs(fshift)
# Convert to valid complex values (imaginary part = 0)
mag_ishift = np.fft.ifftshift(mag_only)
mag_img = np.abs(np.fft.ifft2(mag_ishift))
plt.figure(figsize=(6,6))
plt.imshow(mag_img, cmap='gray')
plt.title("Magnitude-Only Reconstruction")
plt.axis("off")
plt.show()


````

Logarithm is used in magnitude spectrum because the range of FFT magnitudes
is extremely large (often 10^6 difference between low and high frequencies).

Without log, the image would look almost completely black.
Log compresses high values and expands small values,
making details visible in the spectrum.
