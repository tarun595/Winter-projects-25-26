````python
from google.colab import files
uploaded = files.upload()
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image



# get uploaded image name
img_name = list(uploaded.keys())[0]

img = Image.open(img_name)

rgb_img  = np.array(img.convert("RGB"), dtype=np.float32)
gray_img = np.array(img.convert("L"), dtype=np.float32)

# creating Convolutionf fn
def convolve(img, kernel):
    kh, kw = kernel.shape
    pad_h, pad_w = kh // 2, kw // 2

    img_padded = np.pad(
        img,
        ((pad_h, pad_h), (pad_w, pad_w)),
        mode="reflect"
    )

    H, W = img.shape
    out = np.zeros((H, W), dtype=np.float32)

    kernel = np.flipud(np.fliplr(kernel))  # true convolution

    for i in range(H):
        for j in range(W):
            region = img_padded[i:i+kh, j:j+kw]
            out[i, j] = np.sum(region * kernel)

    return out
# average blur
def average_blur_gray(img, ksize):
    kernel = np.ones((ksize, ksize), dtype=np.float32)
    kernel /= kernel.size
    return convolve(img, kernel)

#average blur rgb
def average_blur_rgb(img, ksize):
    out = np.zeros_like(img, dtype=np.float32)
    for c in range(3):
        out[:, :, c] = average_blur_gray(img[:, :, c], ksize)
    return out

#guassian kernel
def gaussian_kernel(ksize, sigma):
    center = ksize // 2
    kernel = np.zeros((ksize, ksize), dtype=np.float32)

    for i in range(ksize):
        for j in range(ksize):
            x = i - center
            y = j - center
            kernel[i, j] = np.exp(-(x*x + y*y) / (2 * sigma * sigma))

    kernel /= np.sum(kernel)
    return kernel

#gaussian blur gray
def gaussian_blur_gray(img, ksize, sigma):
    return convolve(img, gaussian_kernel(ksize, sigma))

# gaussian blur rgb
def gaussian_blur_rgb(img, ksize, sigma):
    out = np.zeros_like(img, dtype=np.float32)
    for c in range(3):
        out[:, :, c] = gaussian_blur_gray(img[:, :, c], ksize, sigma)
    return out
avg_gray   = average_blur_gray(gray_img, 5)
gauss_gray = gaussian_blur_gray(gray_img, 5, 1.0)

avg_rgb   = average_blur_rgb(rgb_img, 5)
gauss_rgb = gaussian_blur_rgb(rgb_img, 5, 1.0)
#results

def show(img, title):
    plt.figure(figsize=(4,4))
    if img.ndim == 2:
        plt.imshow(img, cmap="gray")
    else:
        plt.imshow(img.astype(np.uint8))
    plt.title(title)
    plt.axis("off")
    plt.show()

show(rgb_img, "Original RGB")
show(avg_rgb, "Average Blur RGB")
show(gauss_rgb, "Gaussian Blur RGB")

show(gray_img, "Original Grayscale")
show(avg_gray, "Average Blur Grayscale")
show(gauss_gray, "Gaussian Blur Grayscale")
````

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2

#Laplacian Kernel
laplacian_kernel = np.array([
    [ 1,  1,  1],
    [ 1, -8,  1],
    [ 1,  1,  1]
], dtype=np.float32)

#Laplacian Image
lap = convolve(gray_img, laplacian_kernel)

#Sharpening with alpha = 0.5
alpha = 0.5
sharp_lap = gray_img - alpha * lap

#Sharpening with alpha = 1
alpha_1 = 1.0
sharp_lap_alpha1 = gray_img - alpha_1 * lap

#cv2 Verification
lap_cv2 = cv2.Laplacian(gray_img.astype(np.float32), cv2.CV_32F, ksize=3)
sharp_cv2 = gray_img - lap_cv2

#Gaussian Blur
blurred = gaussian_blur_gray(gray_img, ksize=5, sigma=1.0)

#Mask Extraction
mask = gray_img - blurred

#Sharpen with positive alpha
alpha_usm = 0.7
sharp_usm = gray_img + alpha_usm * mask

#Negative alpha
alpha_negative = -0.5
blurred_via_negative = gray_img + alpha_negative * mask

def show(img, title):
    plt.figure(figsize=(4,4))
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

#  results
show(gray_img, "Original Grayscale")
show(lap, "Laplacian Image")
show(sharp_lap, "Laplacian Sharpening (alpha = 0.5)")
show(sharp_lap_alpha1, "Laplacian Sharpening (alpha = 1)")


show(blurred, "Gaussian Blurred Image")
show(mask, "Unsharp Mask")
show(sharp_usm, "Unsharp Masking (alpha = 0.7)")
show(blurred_via_negative, "Negative Alpha (Blurred Effect)")


````

#### Task - Frequency Analysis (LPF & HPF Verification)
Compute the magnitude spectrum of the original image, gaussian blurred image and the unsharp masking image. Now, play with the alpha value in sharpening and gaussian kernel and notice how the magntude spectrum changes.
Then, in code comments, explain your observations of dark and bright areas in the magnitude spectrum. In nutshell, verify how convolution is actually filtering frequencies depending on the kernel.

````python
import numpy as np
import matplotlib.pyplot as plt

def magnitude_spectrum(img):

    F = np.fft.fft2(img)
    F_shift = np.fft.fftshift(F)
    magnitude = np.log(1 + np.abs(F_shift))
    return magnitude

# Original image
orig = gray_img

# Gaussian blurred image (LPF)
blurred = gaussian_blur_gray(gray_img, ksize=5, sigma=1.0)

# Unsharp masking image (HPF effect)
sharpened = sharp_usm   # from Part C

spec_orig = magnitude_spectrum(orig)
spec_blur = magnitude_spectrum(blurred)
spec_sharp = magnitude_spectrum(sharpened)

def show(img, title):
    plt.figure(figsize=(4,4))
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Spatial domain images
show(orig, "Original Image")
show(blurred, "Gaussian Blurred Image (LPF)")
show(sharpened, "Unsharp Masking Image (HPF)")

# Frequency domain (magnitude spectra)
show(spec_orig, "Magnitude Spectrum: Original Image")
show(spec_blur, "Magnitude Spectrum: Gaussian Blurred Image")
show(spec_sharp, "Magnitude Spectrum: Unsharp Masking Image")

# OBSERVATIONS :


#    - Bright center indicates presence of low-frequency components
#    - Outer regions show high-frequency details (edges)
#

#    - Central region becomes brighter
#    - High-frequency components are suppressed (darker outer region)
#    - Confirms Gaussian blur behaves as a Low-Pass Filter (LPF)


#    - Outer frequency components become brighter
#    - High-frequency details (edges) are amplified
#    - Confirms sharpening behaves as High-Pass Filtering (HPF)

# Thus, convolution in spatial domain corresponds to frequency filtering,



````

````python
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

uploaded = files.upload()
img_name = list(uploaded.keys())[0]

img = Image.open(img_name).convert("RGB")
rgb_img = np.array(img, dtype=np.float32)

def convolve(img, kernel):
    kh, kw = kernel.shape
    pad_h, pad_w = kh // 2, kw // 2

    img_padded = np.pad(
        img,
        ((pad_h, pad_h), (pad_w, pad_w)),
        mode='reflect'
    )

    H, W = img.shape
    out = np.zeros((H, W), dtype=np.float32)

    kernel = np.flipud(np.fliplr(kernel))

    for i in range(H):
        for j in range(W):
            region = img_padded[i:i+kh, j:j+kw]
            out[i, j] = np.sum(region * kernel)

    return out

R = rgb_img[:, :, 0]
G = rgb_img[:, :, 1]
B = rgb_img[:, :, 2]

# Yellow  high R + high G  low B
yellow_response = R + G - 2 * B
yellow_norm = yellow_response - yellow_response.min()
yellow_norm = yellow_norm / yellow_norm.max()
yellow_norm = (yellow_norm * 255).astype(np.uint8)

sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)

sobel_y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
], dtype=np.float32)

Gx = convolve(yellow_norm.astype(np.float32), sobel_x)
Gy = convolve(yellow_norm.astype(np.float32), sobel_y)

yellow_edges = np.sqrt(Gx**2 + Gy**2)

yellow_edges = yellow_edges / yellow_edges.max()
yellow_edges = (yellow_edges * 255).astype(np.uint8)

def show(img, title, cmap=None):
    plt.figure(figsize=(4,4))
    if cmap:
        plt.imshow(img, cmap=cmap)
    else:
        plt.imshow(img)
    plt.title(title)
    plt.axis("off")
    plt.show()

show(rgb_img.astype(np.uint8), "Original Colour Image")
show(yellow_norm, "Yellow Colour Response", cmap="gray")
show(yellow_edges, "Edges of Yellow Leaf", cmap="gray")





````
