````python
# Your PART C codes follow here. You can create more cells under this. Write clean and commented code.
radii = [10,30,80]
h, w = gray_img_cv.shape[:2]

cy, cx = h // 2, w // 2
Y, X = np.ogrid[:h, :w]
dist_from_center = np.sqrt((Y - cy)**2 + (X - cx)**2)

plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.imshow(gray_img_cv, cmap='gray')
plt.title('Original')
plt.axis('off')


for i, r in enumerate(radii):
  mask = np.zeros((h,w), dtype = np.uint8)
  mask[dist_from_center <= r] = 1
  fshift_mask = fshift*mask
  ifshift = np.fft.ifftshift(fshift_mask)
  img_recon = np.fft.ifft2(ifshift)
  img_recon = np.abs(img_recon)

  plt.subplot(1, 4, i + 2)
  plt.imshow(img_recon, cmap='gray')
  plt.title(f'Radius = {r}')
  plt.axis('off')


# smaller radius means fewer low-frequency components pass means stronger blur
````
