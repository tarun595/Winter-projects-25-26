````python
# Creating a LPF mask of radius 10

rows, cols = gray_cv2.shape
crow, ccol = rows//2, cols//2
rad1 = 10
mask_lpf = np.zeros((rows, cols), np.uint8)
cv2.circle(mask_lpf, (ccol, crow), rad1, 1, -1)
f_lpf = fshift * mask_lpf
ishift_lpf = np.fft.ifftshift(f_lpf)
img_back_lpf = np.fft.ifft2(ishift_lpf)
img_back_lpf = np.abs(img_back_lpf)
plt.figure(figsize = (6,6))
plt.imshow(img_back_lpf, cmap = "gray")
plt.title("LPF mask of radius 10")
plt.axis("off")
plt.show()
````

````python
# Creating LPF mask of radius 30

rad2 = 30
mask_lpf = np.zeros((rows, cols), np.uint8)
cv2.circle(mask_lpf, (ccol, crow), rad2, 1, -1)
f_lpf = fshift * mask_lpf
ishift_lpf = np.fft.ifftshift(f_lpf)
img_back_lpf = np.fft.ifft2(ishift_lpf)
img_back_lpf = np.abs(img_back_lpf)
plt.figure(figsize = (6,6))
plt.imshow(img_back_lpf, cmap = "gray")
plt.title("LPF mask of radius 30")
plt.axis("off")
plt.show()
````

````python
# Creating LPF mask of radius 80

rad3 = 80
mask_lpf = np.zeros((rows, cols), np.uint8)
cv2.circle(mask_lpf, (ccol, crow), rad3, 1, -1)
f_lpf = fshift * mask_lpf
ishift_lpf = np.fft.ifftshift(f_lpf)
img_back_lpf = np.fft.ifft2(ishift_lpf)
img_back_lpf = np.abs(img_back_lpf)
plt.figure(figsize = (6,6))
plt.imshow(img_back_lpf, cmap = "gray")
plt.title("LPF mask of radius 80")
plt.axis("off")
plt.show()

# We observe that the blur levels keep increasing on decreasing radius because a smaller radius keeps
# fewer low-frequency components, meaning more high-frequency detail is removed, which produces a stronger blur.
````

````python
# Creating HPF mask of radius 20 (size not mentioned therefore, chosen at random)

rad = 20
mask_lpf = np.zeros((rows, cols), np.uint8)
cv2.circle(mask_lpf, (ccol, crow), rad, 1, -1)
mask_hpf = 1 - mask_lpf
f_hpf = fshift * mask_hpf
ishift_hpf = np.fft.ifftshift(f_hpf)
img_back_hpf = np.fft.ifft2(ishift_hpf)
img_back_hpf = np.abs(img_back_hpf)
plt.figure(figsize = (6,6))
plt.imshow(img_back_hpf, cmap = "gray")
plt.title("HPF mask of radius 20")
plt.axis("off")
plt.show()
````

````python
# Creating a square LPF mask of side 50

rows, cols = gray_cv2.shape
crow, ccol = rows//2, cols//2
side = 50
mask_lpf = np.zeros((rows, cols), np.uint8)
# For creating square at the centre
for i in range(25):
  for j in range(25):
    mask_lpf[crow+i,ccol+j] = 1
    mask_lpf[crow+i,ccol-j] = 1
    mask_lpf[crow-i,ccol-j] = 1
    mask_lpf[crow-i,ccol+j] = 1
f_lpf = fshift * mask_lpf
ishift_lpf = np.fft.ifftshift(f_lpf)
img_back_lpf = np.fft.ifft2(ishift_lpf)
img_back_lpf = np.abs(img_back_lpf)
plt.figure(figsize = (6,6))
plt.imshow(img_back_lpf, cmap = "gray")
plt.title("LPF mask of square of side 50")
plt.axis("off")
plt.show()

# Since the area of this square is greater than that of the circles of radius 10, thus the image appears less blur than that, but its area
# is larger than the circles of radii 30 and 80, that is why it is appearing more blur than them
````

````python
# Phase-only reconstruction (through the given line)

phase_only = fshift / np.abs(fshift)
ishift_phase_only = np.fft.ifftshift(phase_only)
img_back_phase_only = np.fft.ifft2(ishift_phase_only)
img_back_phase_only = np.abs(img_back_phase_only)
plt.figure(figsize = (6,6))
plt.imshow(img_back_phase_only, cmap = "gray")
plt.title("Phase-only reconstruction")
plt.axis("off")
plt.show()

# np.angle() returns raw angles, while the above expression returns exp(j*angle) which is the phase and ifft requires these complex exponents to reconstruct the image but
# since phase only contains structural information, thus without magnitude, only outlines form in the output image
````

````python
# Phase-only reconstruction using np.angle

phase_only_angle = np.angle(fshift)
ishift_phase_only_angle = np.fft.ifftshift(phase_only_angle)
img_back_phase_only_angle = np.fft.ifft2(ishift_phase_only_angle)
img_back_phase_only_angle = np.abs(img_back_phase_only_angle)
plt.figure(figsize = (6,6))
plt.imshow(img_back_phase_only_angle, cmap = "gray")
plt.title("Phase-only reconstruction using np.angle")
plt.axis("off")
plt.show()

# I researched about it and found out that since this passes raw angles in the ifft, thus the expexted strcture that should pass in the ifft is not there anymore and also
# the fourier transform satisfies harmitian symmetry and this output is non-hermitian that is why the ifft produces this, mirrored and distorted image.
````

````python
# Magnitude-oly reconstruction

mag_only = np.abs(fshift)
ishift_mag_only = np.fft.ifftshift(mag_only)
img_back_mag_only = np.fft.ifft2(ishift_mag_only)
img_back_mag_only = np.abs(img_back_mag_only)
plt.figure(figsize = (6,6))
plt.imshow(img_back_mag_only, cmap = "gray")
plt.title("Magnitude-only reconstruction")
plt.axis("off")
plt.show()

# Logarithm does a good amount of scaling to the magnitude i.e., shrinks larger values and expands smaller values making all values somewhat comparable which is not done
# when using only the magnitudes.
````

````python
# Loading the second image

from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

base64_image1 = """"""

img2_data = base64.b64decode(base64_image1)
img2_p= Image.open(io.BytesIO(img2_data))
plt.imshow(img2_p)
plt.axis("off")
print(img2_p.mode)
````

````python
imgrgb_2 = np.array(img2_p)
plt.figure(figsize = (6,6))
plt.imshow(imgrgb_2)
plt.title("Gojo ")
plt.axis("off")
plt.show()
````

````python
# Now applying fft to the R, G, B channels of first image (Joker) and second image(L)

# Joker
img_arr1 = np.array(imgrgb1)
R1 = img_arr1[:,:,0].astype(np.float32)
G1 = img_arr1[:,:,1].astype(np.float32)
B1 = img_arr1[:,:,2].astype(np.float32)

fft_R1 = np.fft.fft2(R1)
fshift_R1 = np.fft.fftshift(fft_R1)
fft_G1 = np.fft.fft2(G1)
fshift_G1 = np.fft.fftshift(fft_G1)
fft_B1 = np.fft.fft2(B1)
fshift_B1 = np.fft.fftshift(fft_B1)

# Gojo
img_arr2 = np.array(imgrgb_2)
R2 = img_arr2[:,:,0].astype(np.float32)
G2 = img_arr2[:,:,1].astype(np.float32)
B2 = img_arr2[:,:,2].astype(np.float32)

fft_R2 = np.fft.fft2(R2)
fshift_R2 = np.fft.fftshift(fft_R2)
fft_G2 = np.fft.fft2(G2)
fshift_G2 = np.fft.fftshift(fft_G2)
fft_B2 = np.fft.fft2(B2)
fshift_B2 = np.fft.fftshift(fft_B2)
````

````python
# Now extractingthe phases

# Joker
phase_R1 = fshift_R1/np.abs((fshift_R1))
phase_G1 = fshift_G1/np.abs((fshift_G1))
phase_B1 = fshift_B1/np.abs((fshift_B1))

# Gojo
phase_R2 = fshift_R2/np.abs((fshift_R2))
phase_G2 = fshift_G2/np.abs((fshift_G2))
phase_B2 = fshift_B2/np.abs((fshift_B2))
````

````python
# Now interchanging the phases

# Joker
new_fshift_R1 = np.abs(fshift_R1)*phase_R2
new_fshift_G1 = np.abs(fshift_G1)*phase_G2
new_fshift_B1 = np.abs(fshift_B1)*phase_B2

# Gojo
new_fshift_R2 = np.abs(fshift_R2)*phase_R1
new_fshift_G2 = np.abs(fshift_G2)*phase_G1
new_fshift_B2 = np.abs(fshift_B2)*phase_B1
````

````python
# Applying ifft and regenerating image of Joker

img1shift_back_R = np.fft.ifft2(np.fft.ifftshift(new_fshift_R1))
img1shift_back_G = np.fft.ifft2(np.fft.ifftshift(new_fshift_G1))
img1shift_back_B = np.fft.ifft2(np.fft.ifftshift(new_fshift_B1))

img1_back_R = np.abs(img1shift_back_R)
img1_back_G = np.abs(img1shift_back_G)
img1_back_B = np.abs(img1shift_back_B)

H, W = img1_back_R.shape
img1_back = np.zeros((H,W,3))

img1_back[:,:,0] = img1_back_R
img1_back[:,:,1] = img1_back_G
img1_back[:,:,2] = img1_back_B

img1_back = np.array(img1_back, dtype = np.uint8)
img1_back = cv2.cvtColor(img1_back, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (6,6))
plt.imshow(img1_back)
plt.title("Joker")
plt.axis("off")
plt.show()
````

````python
# Appplying ifft and regenrating image of Gojo

img2shift_back_R = np.fft.ifft2(np.fft.ifftshift(new_fshift_R2))
img2shift_back_G = np.fft.ifft2(np.fft.ifftshift(new_fshift_G2))
img2shift_back_B = np.fft.ifft2(np.fft.ifftshift(new_fshift_B2))

img2_back_R = np.abs(img2shift_back_R)
img2_back_G = np.abs(img2shift_back_G)
img2_back_B = np.abs(img2shift_back_B)

img2_back = np.zeros((H,W,3))

img2_back[:,:,0] = img2_back_R
img2_back[:,:,1] = img2_back_G
img2_back[:,:,2] = img2_back_B

img2_back = np.array(img2_back, dtype = np.uint8)
img2_back = cv2.cvtColor(img2_back, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (6,6))
plt.imshow(img2_back)
plt.title("Gojo")
plt.axis("off")
plt.show()
````
