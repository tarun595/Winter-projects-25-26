````python


# Define the convolution function from scratch
def my_conv2d(image, kernel):
    # Get image and kernel dimensions
    img_h, img_w = image.shape
    k_h, k_w = kernel.shape

    # Calculate padding (valid for odd-sized kernels)
    pad_h = k_h // 2
    pad_w = k_w // 2

    # Pad the image with zeros or edge values
    padded_img = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='edge')

    # Create empty output image
    output = np.zeros_like(image, dtype=np.float32)

    # Convolution loop
    for i in range(img_h):
        for j in range(img_w):
            # Extract region of interest
            region = padded_img[i:i+k_h, j:j+k_w]
            # Element-wise multiply and sum
            output[i, j] = np.sum(region * kernel)

    # Clip values to 0-255 range
    return np.clip(output, 0, 255)
````

````python
# Create Average Kernel (1/9)
avg_kernel = np.ones((3, 3)) / 9.0

# Apply convolution
blurred_avg = my_conv2d(gray, avg_kernel)

# Display
plt.figure(figsize=(10,5))
plt.subplot(1,2,1); plt.imshow(gray, cmap='gray'); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(blurred_avg, cmap='gray'); plt.title("Average Blur")
plt.show()
````
