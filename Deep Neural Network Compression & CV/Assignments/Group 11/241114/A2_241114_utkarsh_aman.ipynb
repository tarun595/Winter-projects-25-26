````python
#2.rgb hist
def rgb_histogram(rgb_img):
  histogram=np.zeros((3,256),dtype=int)
  for row in rgb_img:
    for pixel in row:
      histogram[0,pixel[0]]+=1
      histogram[1,pixel[1]]+=1
      histogram[2,pixel[2]]+=1
  return histogram

rgb_hist=rgb_histogram(img_np)

plt.plot(rgb_hist[0], color='r')   # Red channel
plt.plot(rgb_hist[1], color='g')   # Green channel
plt.plot(rgb_hist[2], color='b')   # Blue channel

plt.title("3 Channel Histogram")
plt.xlabel("Pixel Intensity (0â€“255)")
plt.ylabel("Frequency")
plt.show()
````

````python
def ForestFreshener(rgb):
    """
    Simple forest filter that:
      1) boosts saturation (make greens pop)
      2) selectively boosts the green channel (channel mixing)
      3) applies slight contrast/brightness adjustment
      4) adds a very subtle vignette to focus the center

    Input: RGB image (uint8, 0..255)
    Output: RGB image (uint8, 0..255)
    """
    # work in float
    img = rgb.astype(np.float32)

    # --- (1) Contrast + brightness (global) ---
    # small contrast increase and slight lift
    alpha = 1.08   # contrast multiplier
    beta  = 6.0    # brightness shift
    img = (img - 128.0) * alpha + 128.0 + beta
    img = np.clip(img, 0, 255)

    # --- (2) Saturation boost via HSV ---
    hsv = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_RGB2HSV).astype(np.float32)
    h, s, v = cv2.split(hsv)
    s = s * 1.30          # increase saturation (makes greens richer)
    s = np.clip(s, 0, 255)
    hsv_boost = cv2.merge([h, s, v]).astype(np.uint8)
    img = cv2.cvtColor(hsv_boost, cv2.COLOR_HSV2RGB).astype(np.float32)

    # --- (3) Green channel boost (channel mixing) ---
    # boost green slightly and slightly reduce red/blue to emphasize foliage
    r = img[...,0] * 0.96
    g = img[...,1] * 1.10
    b = img[...,2] * 0.95
    img = np.stack([r, g, b], axis=-1)
    img = np.clip(img, 0, 255)

    # --- (4) Subtle vignette (optional focus effect) ---
    H, W = img.shape[:2]
    # create distance map from center, normalized to 0..1
    cy, cx = H/2.0, W/2.0
    Y, X = np.ogrid[:H, :W]
    dist = np.sqrt((X - cx)**2 + (Y - cy)**2)
    max_dist = np.sqrt(cx**2 + cy**2)
    vignette = 1.0 - 0.6*(dist / max_dist)   # center ~1.0, edges lower
    vignette = np.clip(vignette, 0.4, 1.0)   # don't go too dark on edges
    vignette = vignette[:, :, None]          # shape (H,W,1)

    img = img * vignette

    out = np.clip(img, 0, 255).astype(np.uint8)
    return out


orig = img2_np.copy()
forest = ForestFreshener(orig)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1); plt.imshow(orig); plt.title("Original"); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(forest); plt.title("ForestFreshener"); plt.axis('off')
plt.show()
````

````python
def white_patch_balance(img_rgb):
    """
    White Patch method:
      - img_rgb: HxWx3 uint8 RGB image (0..255)
      - returns: HxWx3 uint8 RGB (0..255) balanced by white-patch scaling
    """
    img = img_rgb.astype(np.float32)

    # find the brightest pixel by sum of channels (R+G+B)
    sums = img.sum(axis=2)                 # HxW
    idx = np.unravel_index(np.argmax(sums), sums.shape)
    brightest = img[idx[0], idx[1], :]     # shape (3,)

    # If any channel in brightest is 0, avoid dividing by zero
    eps = 1e-9
    factors = 255.0 / (brightest + eps)    # scale per channel

    # apply scaling to whole image
    balanced = img * factors[None, None, :]   # broadcast
    balanced = np.clip(balanced, 0, 255).astype(np.uint8)

    return balanced, brightest.astype(np.uint8), factors

# -----------------------
# Demo / usage
# -----------------------
# try to use img_np from earlier cells, else load fallback
orig = img_np.copy()


balanced, brightest_pixel, scale_factors = white_patch_balance(orig)

# show original and balanced
plt.figure(figsize=(12,5))
plt.subplot(1,3,1); plt.imshow(orig); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(balanced); plt.title("White-Patch Balanced"); plt.axis('off')
# show brightest pixel color as a small patch
patch = np.ones((50,50,3), dtype=np.uint8) * brightest_pixel[None,None,:]
plt.subplot(1,3,3); plt.imshow(patch); plt.title(f"Brightest pixel\n{brightest_pixel.tolist()}"); plt.axis('off')
plt.show()

print("Brightest pixel (R,G,B):", brightest_pixel.tolist())
print("Scale factors (R,G,B):", np.round(scale_factors, 3).tolist())


orig2 = img2_np.copy()
balanced, brightest_pixel, scale_factors = white_patch_balance(orig2)

print(brightest_pixel)
print(scale_factors)

# show original and balanced
plt.figure(figsize=(12,5))
plt.subplot(1,3,1); plt.imshow(orig2); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(balanced); plt.title("White-Patch Balanced"); plt.axis('off')
# show brightest pixel color as a small patch
patch = np.ones((50,50,3), dtype=np.uint8) * brightest_pixel[None,None,:]
plt.subplot(1,3,3); plt.imshow(patch); plt.title(f"Brightest pixel\n{brightest_pixel.tolist()}"); plt.axis('off')
plt.show()



print("Brightest pixel (R,G,B):", brightest_pixel.tolist())
print("Scale factors (R,G,B):", np.round(scale_factors, 3).tolist())
````
