````python
#B1.Computing the FFT of the grayscale image ( use NumPy )
F=np.fft.fft2(gray_np)  #shift zero-frequency to center
fshift=np.fft.fftshift(F)


````

````python
magnitude = np.abs(fshift)     #magnitude
phase = np.angle(fshift)        #later  used in art D

mag_display = np.log(1 + magnitude)   # avoiding log(0) BY +1 AND LOG TO DISPLAY PROPERLY AS MAGNITITUDE SPANS OVER LARGE RANGE

plt.imshow(mag_display, cmap='gray')
plt.title('Magnitude Spectrum (log scale)')
plt.axis('off')
plt.show()
````

````python
#C1 create a mask funtion for circle filter
def lpf_mask(shape, radius):
    H, W = shape
    cy, cx = H // 2, W // 2
    Y, X = np.ogrid[:H, :W]
    dist2 = (Y - cy)**2 + (X - cx)**2
    mask = (dist2 <= radius**2).astype(np.float32)
    return mask

radii = [10, 30, 80]


plt.figure(figsize=(15, 5)) # Create a single figure to hold multiple subplots

for i, r in enumerate(radii):
   #generating mask for each radii
    mask = lpf_mask((H, W), radius=r)
    # Apply mask in frequency domain
    fshift_masked = fshift * mask

    f_ishift = np.fft.ifftshift(fshift_masked) #inverse fshift
    img_back = np.fft.ifft2(f_ishift)

    # Take the real part of the reconstructed image, as the imaginary part should be negligible
    # for real-valued input images after an ideal inverse FFT.
    img_back = np.real(img_back)  # discard tiny imaginary residuals

    # Normalize the pixel values to the 0-255 range and convert to uint8 for display.
    img_back = np.clip(img_back, 0, 255) # Ensure values are within the displayable range
    img_back = img_back.astype(np.uint8) # Convert to unsigned 8-bit integer wrna error ata h

    plt.subplot(1, len(radii), i + 1) # Create subplot for each filtered image (1 row, N columns)
    plt.imshow(img_back, cmap='gray')
    plt.title(f"LPF radius={r}")
    plt.axis('off')

plt.show()

# Reason for blur levels increasing as radius decreases:
# In the frequency domain, the center of the spectrum represents low frequencies (smooth areas),
# while the outer regions represent high frequencies (details, edges, textures).
# A LPF allows low frequencies to pass through and blocks high frequencies.
# When the radius of the LPF mask decreases, it means a smaller circular region around the center of the frequency spectrum is retained. This effectively blocks more high-frequency components.
# Removing more high frequencies results in a smoother image with fewer details and sharper edges,
# which visually translates to increased blurring. Conversely, a larger radius retains more
# high-frequency components, leading to a less blurred image that is closer to the original.



````

````python
# D1. Phase-only reconstruction

# Phase-only spectrum: magnitude = 1, phase preserved
phase_only = fshift / np.abs(fshift)

# Reconstruct from phase-only
img_back_phase = np.fft.ifft2(np.fft.ifftshift(phase_only))
img_back_phase = np.real(img_back_phase)
img_back_phase = np.clip(img_back_phase, 0, 255).astype(np.uint8)

plt.figure(figsize=(6, 6))
plt.imshow(img_back_phase, cmap='gray')
plt.title('Phase-Only Reconstruction')
plt.axis('off')
plt.show()

# Difference between fshift / abs(fshift) and np.angle(fshift):
# - fshift / abs(fshift) produces complex numbers of the form e^(j*phase)
#   which is valid frequency-domain data for inverse FFT.
# - np.angle(fshift) returns only the phase angles as a real array (not complex),
#   so it cannot be used directly for reconstruction. To use angles, we must do exp(1j * angle).

# Attempting inverse FFT on np.angle(fshift) alone gives meaningless output
wrong_recon = np.fft.ifft2(np.fft.ifftshift(np.angle(fshift)))
wrong_recon = np.real(wrong_recon)
wrong_recon = np.clip(wrong_recon - wrong_recon.min(), 0, 255).astype(np.uint8)


plt.figure(figsize=(6,6))
plt.imshow(wrong_recon, cmap='gray')
plt.title('IFFT of np.angle(fshift) (Incorrect Reconstruction)')
plt.axis('off')
plt.show()
````

````python
# D2. Magnitude-only reconstruction

# Magnitude spectrum with zero phase
mag_only_complex = np.abs(fshift) + 0j

# Reconstruct from magnitude-only
img_back_mag = np.fft.ifft2(np.fft.ifftshift(mag_only_complex))
img_back_mag = np.real(img_back_mag)
img_back_mag = np.clip(img_back_mag, 0, 255).astype(np.uint8)

plt.figure(figsize=(6, 6))
plt.imshow(img_back_mag, cmap='gray')
plt.title('Magnitude-Only Reconstruction')
plt.axis('off')
plt.show()

# Why magnitude-only looks different from log magnitude spectrum:
# - The log in the magnitude spectrum (log(1 + |F|)) is ONLY for visualization
#   because raw magnitudes span a huge range.
# - Reconstruction uses the actual magnitudes without log. Combined with zero phase,
#   this loses spatial structure, so the magnitude-only image appears blurred and non-recognizable.
````
