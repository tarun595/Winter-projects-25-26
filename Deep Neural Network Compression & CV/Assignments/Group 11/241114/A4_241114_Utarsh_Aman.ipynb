To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
from PIL import Image
import io, base64
import numpy as np
import matplotlib.pyplot as plt
import cv2

def sobel_analysis(img, title, threshold=None):
    # Step 1: convert RGB image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # Step 2: compute Sobel derivatives
    Ix = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

    # Step 3: prepare gradient values for scatter visualization
    Ix_flat = Ix.reshape(-1)
    Iy_flat = Iy.reshape(-1)

    # Step 4: optional gradient magnitude thresholding
    if threshold is not None:
        valid = (np.abs(Ix_flat) > threshold) | (np.abs(Iy_flat) > threshold)
        Ix_flat = Ix_flat[valid]
        Iy_flat = Iy_flat[valid]

    # Step 5: visualization
    plt.figure(figsize=(16,4))

    plt.subplot(1,4,1)
    plt.imshow(img)
    plt.title(f"{title} : Original")
    plt.axis("off")

    plt.subplot(1,4,2)
    plt.imshow(Ix, cmap="gray")
    plt.title(f"{title} : Ix")
    plt.axis("off")

    plt.subplot(1,4,3)
    plt.imshow(Iy, cmap="gray")
    plt.title(f"{title} : Iy")
    plt.axis("off")

    plt.subplot(1,4,4)
    plt.scatter(Ix_flat, Iy_flat, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(f"{title} : Iy vs Ix")

    plt.tight_layout()
    plt.show()

    return Ix, Iy


# Threshold used to remove noise
THRESHOLD = 20

Ix_flat, Iy_flat = sobel_analysis(flat, "Flat Image", threshold=THRESHOLD)
Ix_edge, Iy_edge = sobel_analysis(edge, "Edge Image", threshold=THRESHOLD)
Ix_corner, Iy_corner = sobel_analysis(corner, "Corner Image", threshold=THRESHOLD)
````

````python
def harris_r_map(img, window_size=3, k=0.04):
    # Convert image to grayscale float format
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY).astype(np.float32)

    # Compute image gradients
    Ix = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

    # Products of derivatives
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy

    # Initialize Harris response matrix
    R = np.zeros(gray.shape)

    offset = window_size // 2

    # Sliding window computation
    for y in range(offset, gray.shape[0] - offset):
        for x in range(offset, gray.shape[1] - offset):

            wx = slice(x - offset, x + offset + 1)
            wy = slice(y - offset, y + offset + 1)

            a = np.sum(Ix2[wy, wx])
            b = np.sum(Ixy[wy, wx])
            c = np.sum(Iy2[wy, wx])

            # Harris corner response
            R[y, x] = (a * c - b * b) - k * (a + c) ** 2

    return R


# Computing R-maps for all 3 images
R_flat   = harris_r_map(flat)
R_edge   = harris_r_map(edge)
R_corner = harris_r_map(corner)


# Displaying all R-maps together
def show_r_map(R_flat, R_edge, R_corner):
    R_flat_n   = cv2.normalize(R_flat,   None, 0, 255, cv2.NORM_MINMAX)
    R_edge_n   = cv2.normalize(R_edge,   None, 0, 255, cv2.NORM_MINMAX)
    R_corner_n = cv2.normalize(R_corner, None, 0, 255, cv2.NORM_MINMAX)

    plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.imshow(R_flat_n, cmap='gray')
    plt.title("Harris R Map - Flat Image")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(R_edge_n, cmap='gray')
    plt.title("Harris R Map - Edge Image")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.imshow(R_corner_n, cmap='gray')
    plt.title("Harris R Map - Corner Image")
    plt.axis("off")

    plt.tight_layout()
    plt.show()


show_r_map(R_flat, R_edge, R_corner)
````

````python
def hough_rho_theta(image,
                    theta_bins=180,
                    rho_bins=250,
                    acc_thresh_ratio=0.9):

    # Preprocess image
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(gray, 100, 200)

    h, w = edges.shape
    diag = int(np.sqrt(h*h + w*w))

    thetas = np.linspace(-np.pi/2, np.pi/2, theta_bins)
    rhos = np.linspace(-diag, diag, rho_bins)

    accumulator = np.zeros((rho_bins, theta_bins), dtype=np.int32)

    ys, xs = np.nonzero(edges)

    # Voting in (rho, theta) space
    for x, y in zip(xs, ys):
        for t_idx, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            r_idx = int((rho + diag) / (2 * diag) * (rho_bins - 1))
            accumulator[r_idx, t_idx] += 1

    # Normalize and threshold
    acc_norm = accumulator / accumulator.max()
    idxs = np.where(acc_norm > acc_thresh_ratio)

    img_lines = image.copy()

    # Draw detected lines
    for r_idx, t_idx in zip(idxs[0], idxs[1]):
        rho = rhos[r_idx]
        theta = thetas[t_idx]

        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a * rho, b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(img_lines, (x1, y1), (x2, y2), (0, 255, 0), 2)

    return edges, accumulator, img_lines


edges_rt, acc_rt, result_rt = hough_rho_theta(ball)

# Side-by-side visualization
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(ball)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(result_rt)
plt.title("RhoTheta Hough Lines")
plt.axis("off")

plt.tight_layout()
plt.show()

edges_rt, acc_rt, result_rt = hough_rho_theta(edge)

# Side-by-side visualization
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(edge)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(result_rt)
plt.title("RhoTheta Hough Lines")
plt.axis("off")

plt.tight_layout()
plt.show()
````

````python
def extract_ball_stencil(image, x1, y1, x2, y2):
    # Crop region
    crop = image[y1:y2, x1:x2]

    gray = cv2.cvtColor(crop, cv2.COLOR_RGB2GRAY)

    # Reduce texture noise -> smoother gradients
    gray = cv2.GaussianBlur(gray, (5,5), 0)

    # Gradient magnitude
    Ix = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
    mag = np.sqrt(Ix**2 + Iy**2)
    mag = (mag / (mag.max() ) * 255).astype(np.uint8)

    # Threshold + closing to complete circular edges
    _, bw = cv2.threshold(mag, 40, 255, cv2.THRESH_BINARY)
    kernel = np.ones((5,5), np.uint8)
    bw = cv2.morphologyEx(bw, cv2.MORPH_CLOSE, kernel)

    # Find contours (safety: handle empty result)
    contours_info = cv2.findContours(bw, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours = contours_info[0] if len(contours_info) == 2 else contours_info[1]

    best = None
    best_score = 0
    min_area = max(200, 0.01 * crop.shape[0] * crop.shape[1])  # ignore tiny noise

    if contours:
        for cnt in contours:
            area = cv2.contourArea(cnt)
            if area < min_area:
                continue
            perim = cv2.arcLength(cnt, True)
            if perim <= 0:
                continue
            circularity = 4 * np.pi * area / (perim * perim)
            score = area * circularity   # prefers large, circular shapes
            if score > best_score:
                best_score = score
                best = cnt


    mask = np.zeros(gray.shape, dtype=np.uint8)
    if best is not None:
        cv2.drawContours(mask, [best], -1, 255, -1)
    elif contours:
        largest = max(contours, key=cv2.contourArea)
        cv2.drawContours(mask, [largest], -1, 255, -1)
    else:
        # fallback: conservative circular mask in center of crop
        cy, cx = crop.shape[0]//2, crop.shape[1]//2
        r = min(cy, cx, 50)
        cv2.circle(mask, (cx, cy), r, 255, -1)

    # Slightly blur alpha to smooth edges (anti-alias)
    mask = cv2.GaussianBlur(mask, (7,7), 0)
    mask = np.clip(mask, 0, 255).astype(np.uint8)

    # RGBA stencil
    rgba = np.zeros((crop.shape[0], crop.shape[1], 4), dtype=np.uint8)
    rgba[..., :3] = crop
    rgba[..., 3] = mask

    return rgba


def overlay_stencil(bg, stencil, x3, y3):
    h, w = stencil.shape[:2]
    x_start = int(x3 - w // 2)
    y_start = int(y3 - h // 2)
    x_end = x_start + w
    y_end = y_start + h

    output = bg.copy()

    # Clip to background bounds and compute corresponding stencil slice
    bx1 = max(0, x_start)
    by1 = max(0, y_start)
    bx2 = min(bg.shape[1], x_end)
    by2 = min(bg.shape[0], y_end)

    if bx1 >= bx2 or by1 >= by2:
        # stencil is completely outside background
        return output

    sx1 = bx1 - x_start
    sy1 = by1 - y_start
    sx2 = sx1 + (bx2 - bx1)
    sy2 = sy1 + (by2 - by1)

    bg_roi = output[by1:by2, bx1:bx2].astype(np.float32)
    st_roi = stencil[sy1:sy2, sx1:sx2, :3].astype(np.float32)
    alpha = stencil[sy1:sy2, sx1:sx2, 3].astype(np.float32) / 255.0

    # If alpha is entirely zero, nothing to do
    if np.all(alpha <= 1e-6):
        return output

    # Expand alpha to 3 channels
    alpha_3 = np.expand_dims(alpha, axis=2)

    # Alpha blend (vectorized)
    composited = (1.0 - alpha_3) * bg_roi + alpha_3 * st_roi
    output[by1:by2, bx1:bx2] = np.clip(composited, 0, 255).astype(np.uint8)

    return output
````
