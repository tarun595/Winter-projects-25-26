```python
# PART A
def gray_histogram(gray_img):
    h, w = gray_img.shape
    hist = np.zeros(256, dtype=np.int64)
    for i in range(h):
        row = gray_img[i]
        for j in range(w):
            hist[int(row[j])] += 1
    return hist

gray = (0.299 * img_rgb[:,:,0] + 0.587 * img_rgb[:,:,1] + 0.114 * img_rgb[:,:,2]).astype(np.uint8)
hist_gray = gray_histogram(gray)

plt.figure()
plt.plot(hist_gray)
plt.title("Grayscale Histogram")
plt.xlabel("Intensity (0-255)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

def rgb_histogram(rgb_img):
    h, w, _ = rgb_img.shape
    hr = np.zeros(256, dtype=np.int64)
    hg = np.zeros(256, dtype=np.int64)
    hb = np.zeros(256, dtype=np.int64)
    for i in range(h):
        for j in range(w):
            r, g, b = rgb_img[i, j]
            hr[int(r)] += 1
            hg[int(g)] += 1
            hb[int(b)] += 1
    return hr, hg, hb

hr, hg, hb = rgb_histogram(img_rgb)
x = np.arange(256)

plt.figure(figsize=(8,3))
plt.plot(hr)
plt.plot(hg)
plt.plot(hb)
plt.title("RGB Histograms (R,G,B)")
plt.tight_layout()
plt.show()

```

```python
# PART B
def rgb_to_hsv(rgb_img):
    arr = rgb_img.astype(np.float32) / 255.0
    R = arr[:,:,0]
    G = arr[:,:,1]
    B = arr[:,:,2]
    Cmax = np.maximum(np.maximum(R,G), B)
    Cmin = np.minimum(np.minimum(R,G), B)
    delta = Cmax - Cmin

    H = np.zeros_like(Cmax)
    nz = delta > 1e-12

    mask = (Cmax == R) & nz
    H[mask] = (60.0 * ((G[mask] - B[mask]) / delta[mask])) % 360.0
    mask = (Cmax == G) & nz
    H[mask] = (60.0 * ((B[mask] - R[mask]) / delta[mask]) + 120.0) % 360.0
    mask = (Cmax == B) & nz
    H[mask] = (60.0 * ((R[mask] - G[mask]) / delta[mask]) + 240.0) % 360.0

    S = np.zeros_like(Cmax)
    nonzero = Cmax > 1e-12
    S[nonzero] = delta[nonzero] / Cmax[nonzero]
    V = Cmax

    H = np.floor(H / 2.0).astype(np.uint8)
    S = np.clip(S * 255.0, 0, 255).astype(np.uint8)
    V = np.clip(V * 255.0, 0, 255).astype(np.uint8)

    return np.stack([H, S, V], axis=2)

def hsv_to_rgb(hsv_img):
    H = hsv_img[:,:,0].astype(np.float32) * 2.0
    S = hsv_img[:,:,1].astype(np.float32) / 255.0
    V = hsv_img[:,:,2].astype(np.float32) / 255.0

    C = V * S
    H_ = H / 60.0
    X = C * (1 - np.abs((H_ % 2) - 1))
    m = V - C

    R = np.zeros_like(H)
    G = np.zeros_like(H)
    B = np.zeros_like(H)

    c0 = (0 <= H_) & (H_ < 1)
    c1 = (1 <= H_) & (H_ < 2)
    c2 = (2 <= H_) & (H_ < 3)
    c3 = (3 <= H_) & (H_ < 4)
    c4 = (4 <= H_) & (H_ < 5)
    c5 = (5 <= H_) & (H_ < 6)

    R[c0] = C[c0]; G[c0] = X[c0]
    R[c1] = X[c1]; G[c1] = C[c1]
    G[c2] = C[c2]; B[c2] = X[c2]
    G[c3] = X[c3]; B[c3] = C[c3]
    R[c4] = X[c4]; B[c4] = C[c4]
    R[c5] = C[c5]; B[c5] = X[c5]

    R = (R + m) * 255.0
    G = (G + m) * 255.0
    B = (B + m) * 255.0

    return np.stack([R, G, B], axis=2).clip(0,255).astype(np.uint8)

hsv_img = rgb_to_hsv(img_rgb)
rgb_recon = hsv_to_rgb(hsv_img)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); imshow_rgb(img_rgb)
plt.subplot(1,2,2); imshow_rgb(rgb_recon)
plt.show()

# Gaussian Blur
def gaussian_kernel1d(ksize, sigma):
    half = ksize // 2
    x = np.arange(-half, half+1, dtype=np.float32)
    g = np.exp(-(x**2) / (2*sigma*sigma))
    g /= g.sum()
    return g

def gaussian_blur(img, ksize=7, sigma=2.0):
    kernel = gaussian_kernel1d(ksize, sigma)
    pad = ksize // 2
    H, W, C = img.shape
    arr = img.astype(np.float32)
    horiz = np.zeros_like(arr)
    for ch in range(C):
        padded = np.pad(arr[:,:,ch], ((0,0),(pad,pad)), mode='reflect')
        for i in range(H):
            horiz[i,:,ch] = np.convolve(padded[i], kernel, mode='valid')
    out = np.zeros_like(arr)
    for ch in range(C):
        padded = np.pad(horiz[:,:,ch], ((pad,pad),(0,0)), mode='reflect')
        for j in range(W):
            out[:,j,ch] = np.convolve(padded[:,j], kernel, mode='valid')
    return np.clip(out,0,255).astype(np.uint8)

```

```python
# PART C Filters
def tropical_punch(img):
    hsv = rgb_to_hsv(img)
    H = hsv[:,:,0].astype(np.float32)
    S = hsv[:,:,1].astype(np.float32)
    V = hsv[:,:,2].astype(np.float32)

    S = np.clip(S * 1.25, 0, 255)
    V = np.clip(V * 1.08, 0, 255)

    hsv2 = np.stack([H.astype(np.uint8), S.astype(np.uint8), V.astype(np.uint8)], axis=2)
    rgb = hsv_to_rgb(hsv2).astype(np.float32)

    r = rgb[:,:,0]; g = rgb[:,:,1]; b = rgb[:,:,2]
    r = np.clip(r * 1.06 + 5, 0, 255)
    b = np.clip(b * 0.95, 0, 255)

    gamma = 0.95
    r = ((r/255.0)**gamma) * 255
    g = ((g/255.0)**gamma) * 255
    b = ((b/255.0)**gamma) * 255

    return np.stack([r,g,b], axis=2).clip(0,255).astype(np.uint8)

def misty_morning(img):
    imgf = img.astype(np.float32)
    img_cb = np.clip((imgf - 128)*0.92 + 128 + 6, 0, 255)

    hsv = rgb_to_hsv(img_cb.astype(np.uint8))
    H = hsv[:,:,0].astype(np.float32)
    S = hsv[:,:,1].astype(np.float32)
    V = hsv[:,:,2].astype(np.float32)
    S = np.clip(S * 0.85, 0, 255)

    hsv2 = np.stack([H.astype(np.uint8), S.astype(np.uint8), V.astype(np.uint8)], axis=2)
    rgb = hsv_to_rgb(hsv2).astype(np.float32)

    Vf = V / 255.0
    hl = np.clip((Vf - 0.6)/0.4, 0, 1)
    sh = np.clip((0.5 - Vf)/0.5, 0, 1)

    rgb[:,:,2] = np.clip(rgb[:,:,2] + hl*12 + sh*6, 0, 255)
    rgb[:,:,0] = np.clip(rgb[:,:,0] - sh*8, 0, 255)

    blurred = gaussian_blur(rgb.astype(np.uint8), 7, 2).astype(np.float32)
    out = 0.75*rgb + 0.25*blurred

    return out.clip(0,255).astype(np.uint8)

trop = tropical_punch(img_rgb)
fog = misty_morning(img_rgb)

plt.figure(figsize=(15,6))
plt.subplot(1,3,1); imshow_rgb(img_rgb, "Original")
plt.subplot(1,3,2); imshow_rgb(trop, "Tropical Punch")
plt.subplot(1,3,3); imshow_rgb(fog, "Misty Morning")
plt.show()
```

```python
# PART D
def white_patch_balance(img):
    f = img.astype(np.float32)
    mr = f[:,:,0].max()
    mg = f[:,:,1].max()
    mb = f[:,:,2].max()
    out = f.copy()
    if mr: out[:,:,0] *= 255.0/mr
    if mg: out[:,:,1] *= 255.0/mg
    if mb: out[:,:,2] *= 255.0/mb
    return out.clip(0,255).astype(np.uint8)

wb = white_patch_balance(img_rgb)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); imshow_rgb(img_rgb, "Original")
plt.subplot(1,2,2); imshow_rgb(wb, "White Patch")
plt.show()
```

```python
# PART E 
def white_balance(img, value):
    t = (value - 0.5) * 2.0
    cold = np.array([0.95,1.00,1.12])
    warm = np.array([1.12,1.00,0.92])
    scale = cold*(1-value) + warm*value

    out = img.astype(np.float32) * scale.reshape(1,1,3)

    gamma = 1.0 - (value - 0.5)*0.08
    out = ((out/255.0)**gamma)*255.0

    hsv = rgb_to_hsv(out.clip(0,255).astype(np.uint8))
    H = hsv[:,:,0].astype(np.float32)
    S = hsv[:,:,1].astype(np.float32)
    V = hsv[:,:,2].astype(np.float32)

    S = np.clip(S * (1 + 0.15*abs(t)), 0, 255)

    hsv2 = np.stack([H.astype(np.uint8), S.astype(np.uint8), V.astype(np.uint8)], axis=2)
    return hsv_to_rgb(hsv2)

vals = [0.0,0.25,0.5,0.75,1.0]
res = [white_balance(img_rgb, v) for v in vals]

plt.figure(figsize=(15,8))
for i,v in enumerate(vals):
    plt.subplot(2,3,i+1)
    imshow_rgb(res[i], f"WB={v}")
plt.subplot(2,3,6)
imshow_rgb(img_rgb, "Original")
plt.show()
```
