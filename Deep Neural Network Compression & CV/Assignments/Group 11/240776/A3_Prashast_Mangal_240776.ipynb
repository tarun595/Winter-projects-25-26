```python
from PIL import Image
import io
import base64
import numpy as np
import matplotlib.pyplot as plt
base64_image = """"""

img_bytes = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_bytes)).convert("RGB")

img_np = np.array(img).astype(np.float32)

plt.imshow(img_np.astype(np.uint8))
plt.axis("off")
plt.title("Original image")
plt.show()
```

```python
#----SECTION 2 : PART A : Task 1-----

def convolve(img, kernel):
    kh, kw = kernel.shape
    pad_h, pad_w = kh // 2, kw // 2

    padded = np.pad(img,
                    ((pad_h, pad_h), (pad_w, pad_w)),
                    mode="reflect")

    out = np.zeros_like(img)

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            region = padded[i:i+kh, j:j+kw]
            out[i, j] = np.sum(region * kernel)

    return out
```

```python
#----SECTION 2 : PART A : Task 2-----
def rgb_to_gray(img):
    return (0.299 * img[...,0] +
            0.587 * img[...,1] +
            0.114 * img[...,2])
def average_blur_gray(img, ksize):
    kernel = np.ones((ksize, ksize), dtype=np.float32)
    kernel /= kernel.size
    return convolve(img, kernel)
def average_blur_rgb(img, ksize):
    out = np.zeros_like(img)
    for c in range(3):
        out[..., c] = average_blur_gray(img[..., c], ksize)
    return out
gray = rgb_to_gray(img_np)
blur_gray = average_blur_gray(gray, 5)
blur_rgb = average_blur_rgb(img_np, 5)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(blur_gray, cmap="gray")
plt.title("Average Blur (Gray)")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(blur_rgb.astype(np.uint8))
plt.title("Average Blur (RGB)")
plt.axis("off")
plt.show()
```

```python
#----SECTION 2 : PART A : Task 3-----
def gaussian_kernel(ksize, sigma):
    ax = np.arange(-(ksize//2), ksize//2 + 1)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
    kernel /= np.sum(kernel)
    return kernel
gk = gaussian_kernel(5, 1.2)

gauss_gray = convolve(gray, gk)

gauss_rgb = np.zeros_like(img_np)
for c in range(3):
    gauss_rgb[..., c] = convolve(img_np[..., c], gk)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(img_np.astype(np.uint8))
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(gauss_gray, cmap="gray")
plt.title("Gaussian (Gray)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(gauss_rgb.astype(np.uint8))
plt.title("Gaussian (RGB)")
plt.axis("off")
plt.show()
```

```python
#-------SECTION 2 : PART B-------
sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]], dtype=np.float32)

sobel_y = np.array([[ 1, 2, 1],
                    [ 0, 0, 0],
                    [-1,-2,-1]], dtype=np.float32)

Gx = convolve(gray, sobel_x)
Gy = convolve(gray, sobel_y)

grad_mag = np.sqrt(Gx**2 + Gy**2)
grad_mag = (grad_mag / grad_mag.max()) * 255
plt.imshow(grad_mag, cmap="gray")
plt.title("Sobel Edge Map")
plt.axis("off")
plt.show()
```

```python
#-------SECTION 2 : PART C : TASK 1-------
laplacian_kernel = np.array([[0,-1,0],
                             [-1,4,-1],
                             [0,-1,0]], dtype=np.float32)

lap = convolve(gray, laplacian_kernel)

alpha = 1.0
sharp_lap = gray - alpha * lap
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(gray, cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sharp_lap, cmap="gray")
plt.title("Laplacian Sharpened")
plt.axis("off")
plt.show()
```

```python
#-------SECTION 2 : PART C : TASK 2-------
blur = convolve(gray, gaussian_kernel(5, 1.0))
mask = gray - blur

alpha = 1.5
unsharp = gray + alpha * mask
plt.imshow(unsharp, cmap="gray")
plt.title("Unsharp Masking")
plt.axis("off")
plt.show()
# Negative alpha  blur
blur_again = gray + (-1.0) * mask
plt.imshow(blur_again, cmap="gray")
plt.title("Negative Alpha (Blur)")
plt.axis("off")
plt.show()
```

```python
#-------SECTION 2 : PART D ------
def magnitude_spectrum(img):
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    mag = np.log(np.abs(fshift) + 1)
    return mag
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(magnitude_spectrum(gray), cmap="gray")
plt.title("Original Spectrum")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(magnitude_spectrum(blur), cmap="gray")
plt.title("Gaussian Blur Spectrum")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(magnitude_spectrum(unsharp), cmap="gray")
plt.title("Sharpened Spectrum")
plt.axis("off")
plt.show()
```

```python
#-------SECTION 2 : PART E -------
R = img_np[...,0]
G = img_np[...,1]
B = img_np[...,2]

# Yellow  R high, G high, B low
rg_diff = R - G

edge_kernel = np.array([[-1,-1,-1],
                        [ 0, 0, 0],
                        [ 1, 1, 1]], dtype=np.float32)

edge_rg = convolve(rg_diff, edge_kernel)
edge_r = convolve(R, edge_kernel)
edge_g = convolve(G, edge_kernel)

combined = np.abs(edge_rg) + np.abs(edge_r - edge_g)
combined = (combined / combined.max()) * 255
plt.imshow(combined, cmap="gray")
plt.title("Yellow Edge Response")
plt.axis("off")
plt.show()
```

```python
#-------SECTION 2 : PART F -------
def rgb_to_hsv_np(img):
    img = img / 255.0
    r, g, b = img[...,0], img[...,1], img[...,2]

    cmax = np.max(img, axis=2)
    cmin = np.min(img, axis=2)
    diff = cmax - cmin

    h = np.zeros_like(cmax)
    s = np.zeros_like(cmax)
    v = cmax

    # Hue computation (safe division)
    mask = diff != 0

    h_r = np.divide((g - b), diff, where=mask)
    h_g = np.divide((b - r), diff, where=mask)
    h_b = np.divide((r - g), diff, where=mask)

    h[(cmax == r) & mask] = (60 * h_r[(cmax == r) & mask]) % 360
    h[(cmax == g) & mask] = (60 * h_g[(cmax == g) & mask]) + 120
    h[(cmax == b) & mask] = (60 * h_b[(cmax == b) & mask]) + 240

    # Saturation
    s[cmax != 0] = diff[cmax != 0] / cmax[cmax != 0]

    return h, s, v

h, s, v = rgb_to_hsv_np(img_np)

# Yellow range 
yellow_mask = (
    (h >= 20) & (h <= 35) &   # hue range
    (s >= 0.4) &              # saturation
    (v >= 0.4)                # brightness
)
# Hue = 0 corresponds to red
h[yellow_mask] = 0
def hsv_to_rgb_np(h, s, v):
    c = v * s
    x = c * (1 - np.abs((h / 60) % 2 - 1))
    m = v - c

    z = np.zeros_like(c)
    rgb = np.zeros((h.shape[0], h.shape[1], 3))

    conds = [
        (0 <= h) & (h < 60),
        (60 <= h) & (h < 120),
        (120 <= h) & (h < 180),
        (180 <= h) & (h < 240),
        (240 <= h) & (h < 300),
        (300 <= h) & (h < 360)
    ]

    values = [
        (c, x, z),
        (x, c, z),
        (z, c, x),
        (z, x, c),
        (x, z, c),
        (c, z, x)
    ]

    for cond, (rc, gc, bc) in zip(conds, values):
        rgb[...,0][cond] = rc[cond]
        rgb[...,1][cond] = gc[cond]
        rgb[...,2][cond] = bc[cond]

    rgb += m[..., None]
    return (rgb * 255).clip(0, 255)

result = hsv_to_rgb_np(h, s, v)

plt.imshow(result.astype(np.uint8))
plt.axis("off")
plt.title("Yellow  Red")
plt.show()

```
