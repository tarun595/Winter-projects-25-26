To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def harris_r_map(image, window_size=3, k=0.04):
    image = image.astype(np.float32)

    Ix = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix*Iy

    Sxx = cv2.GaussianBlur(Ixx, (window_size,window_size), 1)
    Syy = cv2.GaussianBlur(Iyy, (window_size,window_size), 1)
    Sxy = cv2.GaussianBlur(Ixy, (window_size,window_size), 1)

    det = (Sxx * Syy) - (Sxy ** 2)
    trace = Sxx + Syy

    R = det - k * (trace ** 2)
    return R
def show_r_map(R, title):
    plt.figure(figsize=(5,5))
    plt.imshow(normalize(R), cmap='hot')
    plt.colorbar()
    plt.title(title)
    plt.axis("off")
    plt.show()
show_r_map(harris_r_map(flat), "R Map - Flat")
show_r_map(harris_r_map(edge), "R Map - Edge")
show_r_map(harris_r_map(corner), "R Map - Corner")
````

````python
def hough_slope_intercept(image, theta_res=1, b_res=2):
    edges = cv2.Canny(image, 50, 150)
    h, w = edges.shape

    thetas = np.deg2rad(np.arange(-90, 90, theta_res))
    bs = np.arange(-h, h, b_res)

    accumulator = np.zeros((len(thetas), len(bs)))

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for ti, theta in enumerate(thetas):
            if abs(np.sin(theta)) < 1e-6:
                continue
            b = y - x * np.cos(theta) / np.sin(theta)
            bi = int((b + h) / b_res)
            if 0 <= bi < len(bs):
                accumulator[ti, bi] += 1

    return accumulator, thetas, bs
def draw_slope_intercept_lines(image, accumulator, thetas, bs, threshold_ratio=0.9):

    if len(image.shape) == 2:
        img = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    else:
        img = image.copy()

    acc_norm = normalize(accumulator)
    t_idxs, b_idxs = np.where(acc_norm > threshold_ratio)

    for t_idx, b_idx in zip(t_idxs, b_idxs):
        theta = thetas[t_idx]
        b = bs[b_idx]

        if abs(np.sin(theta)) < 1e-6:
            continue

        x0, x1 = 0, image.shape[1]
        y0 = int((x0 * np.cos(theta) / np.sin(theta)) + b)
        y1 = int((x1 * np.cos(theta) / np.sin(theta)) + b)

        cv2.line(img, (x0, y0), (x1, y1), (255, 0, 0), 1)

    show(img, "SlopeIntercept Hough Output")
accumulator, thetas, bs = hough_slope_intercept(hough_img)
draw_slope_intercept_lines(hough_img, accumulator, thetas, bs)
````

````python
import cv2
import numpy as np

def create_ball_stencil_from_gradients(image, x1, y1, x2, y2):
    """
    image : grayscale image containing the white ball
    (x1, y1) : top-left of bounding box
    (x2, y2) : bottom-right of bounding box

    Returns:
        RGBA stencil (H x W x 4)
    """


    roi = image[y1:y2, x1:x2]

    # --- Compute gradients ---
    gx = cv2.Sobel(roi, cv2.CV_64F, 1, 0, ksize=3)
    gy = cv2.Sobel(roi, cv2.CV_64F, 0, 1, ksize=3)
    grad_mag = np.sqrt(gx**2 + gy**2)

    # --- Normalize gradient magnitude ---
    grad_mag = grad_mag - grad_mag.min()
    grad_mag = grad_mag / (grad_mag.max() + 1e-8)

    # --- Threshold gradients to get edges ---
    edge_mask = (grad_mag > 0.2).astype(np.uint8) * 255

    # --- Close gaps to form completed contours ---
    kernel = np.ones((5,5), np.uint8)
    edge_mask = cv2.morphologyEx(edge_mask, cv2.MORPH_CLOSE, kernel)

    # --- Find contours ---
    contours, _ = cv2.findContours(edge_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) == 0:
        raise ValueError("No contours found in gradient image.")

    # --- Select largest contour (maximum area) ---
    largest_contour = max(contours, key=cv2.contourArea)

    # --- Create alpha mask ---
    alpha = np.zeros_like(roi, dtype=np.uint8)
    cv2.drawContours(alpha, [largest_contour], -1, 255, -1)

    # --- Create RGBA stencil ---
    rgba = cv2.cvtColor(roi, cv2.COLOR_GRAY2BGRA)
    rgba[:, :, 3] = alpha

    return rgba
def overlay_stencil_on_image(background, stencil, x3, y3):
    """
    background : grayscale image (target image)
    stencil    : RGBA image from create_ball_stencil_from_gradients
    (x3, y3)   : center position where stencil will be placed

    Returns:
        BGR image with stencil overlaid
    """

    # Convert background to BGR for visualization
    bg = cv2.cvtColor(background, cv2.COLOR_GRAY2BGR)

    h, w = stencil.shape[:2]

    # Compute placement coordinates
    x1 = int(x3 - w // 2)
    y1 = int(y3 - h // 2)
    x2 = x1 + w
    y2 = y1 + h

    # Clip to image boundaries
    x1_clip, y1_clip = max(0, x1), max(0, y1)
    x2_clip, y2_clip = min(bg.shape[1], x2), min(bg.shape[0], y2)

    stencil = stencil[(y1_clip - y1):(y2_clip - y1),
                      (x1_clip - x1):(x2_clip - x1)]

    alpha = stencil[:, :, 3] / 255.0

    # Alpha blending
    for c in range(3):
        bg[y1_clip:y2_clip, x1_clip:x2_clip, c] = (
            alpha * stencil[:, :, c] +
            (1 - alpha) * bg[y1_clip:y2_clip, x1_clip:x2_clip, c]
        )

    return bg
# Ensure grayscale
ball_gray = ball if len(ball.shape) == 2 else cv2.cvtColor(ball, cv2.COLOR_BGR2GRAY)
ball2_gray = ball2 if len(ball2.shape) == 2 else cv2.cvtColor(ball2, cv2.COLOR_BGR2GRAY)

# Manually chosen bounding box around ball (adjust if needed)
x1, y1, x2, y2 = 80, 80, 200, 200

stencil = create_ball_stencil_from_gradients(ball_gray, x1, y1, x2, y2)

# Place ball towards the right side of second image
final_image = overlay_stencil_on_image(ball2_gray, stencil, x3=220, y3=150)
````
