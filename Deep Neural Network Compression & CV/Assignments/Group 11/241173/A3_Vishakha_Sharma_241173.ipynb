````python
##Task 1

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img_rgb = Image.open("/content/kanhaji.jpg").convert("RGB")
img_rgb = np.array(img_rgb)

img_gray = Image.open("/content/kanhaji.jpg").convert("L")
img_gray = np.array(img_gray)


def convolve(img, kernel):
    k = kernel.shape[0]
    pad = k // 2

    padded = np.pad(img, pad, mode='reflect')
    h, w = img.shape

    out = np.zeros((h, w))

    for i in range(h):
        for j in range(w):
            region = padded[i:i+k, j:j+k]
            out[i, j] = np.sum(region * kernel)

    return out

##Task 2

def average_blur_gray(img, ksize):
    kernel = np.ones((ksize, ksize))
    kernel = kernel / (ksize * ksize)
    return convolve(img, kernel)

avg_gray = average_blur_gray(img_gray, 3)
plt.imshow(avg_gray, cmap='gray')
plt.axis('off')
plt.show()

def average_blur_rgb(img, ksize):
    out = np.zeros_like(img, dtype=float)

    for c in range(3):
        out[:, :, c] = average_blur_gray(img[:, :, c], ksize)

    return out
avg_rgb = average_blur_rgb(img_rgb, 3)
plt.imshow(avg_rgb.astype(np.uint8))
plt.axis('off')
plt.show()

````

````python
##Task 3
def gaussian_kernel(ksize, sigma):
    k = ksize // 2
    kernel = np.zeros((ksize, ksize))

    for i in range(ksize):
        for j in range(ksize):
            x = i - k
            y = j - k
            kernel[i, j] = np.exp(-(x*x + y*y) / (2 * sigma * sigma))

    kernel = kernel / kernel.sum()
    return kernel

def gaussian_blur_gray(img, ksize, sigma):
    kernel = gaussian_kernel(ksize, sigma)
    return convolve(img, kernel)
gauss_gray = gaussian_blur_gray(img_gray, 9, 1.0)
plt.imshow(gauss_gray, cmap='gray')
plt.axis('off')
plt.show()

def gaussian_blur_rgb(img, ksize, sigma):
    out = np.zeros_like(img, dtype=float)

    for c in range(3):
        out[:, :, c] = gaussian_blur_gray(img[:, :, c], ksize, sigma)

    return out
gauss_rgb = gaussian_blur_rgb(img_rgb, 9, 1.0)
plt.imshow(gauss_rgb.astype(np.uint8))
plt.axis('off')
plt.show()
````

#### Task - Frequency Analysis (LPF & HPF Verification)
Compute the magnitude spectrum of the original image, gaussian blurred image and the unsharp masking image. Now, play with the alpha value in sharpening and gaussian kernel and notice how the magntude spectrum changes.
Then, in code comments, explain your observations of dark and bright areas in the magnitude spectrum. In nutshell, verify how convolution is actually filtering frequencies depending on the kernel.

````python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open("/content/kanhaji.jpg").convert("L")
img = np.array(img, dtype=float)

##convolution
def convolve(img, kernel):
    k = kernel.shape[0]
    pad = k // 2

    padded = np.pad(img, pad, mode='reflect')
    h, w = img.shape
    out = np.zeros((h, w))

    for i in range(h):
        for j in range(w):
            region = padded[i:i+k, j:j+k]
            out[i, j] = np.sum(region * kernel)

    return out

##gaussian kernel
def gaussian_kernel(ksize, sigma):
    k = ksize // 2
    kernel = np.zeros((ksize, ksize))

    for i in range(ksize):
        for j in range(ksize):
            x = i - k
            y = j - k
            kernel[i, j] = np.exp(-(x*x + y*y) / (2 * sigma * sigma))

    kernel = kernel / kernel.sum()
    return kernel

##unsharp masking
kernel = gaussian_kernel(5, 1.0)
blurred = convolve(img, kernel)

mask = img - blurred
alpha = 1.0
sharp = img + alpha * mask

##magnitude spectrum
def magnitude_spectrum(img):
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    mag = np.log(1 + np.abs(fshift))
    return mag
##computing spectra
spec_orig = magnitude_spectrum(img)
spec_blur = magnitude_spectrum(blurred)
spec_sharp = magnitude_spectrum(sharp)

##display
plt.figure(figsize=(9,3))

plt.subplot(1,3,1)
plt.imshow(spec_orig, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(spec_blur, cmap='gray')
plt.title("Gaussian Blur")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(spec_sharp, cmap='gray')
plt.title("Unsharp Masking")
plt.axis('off')

plt.show()

# Observations:
# The center of the magnitude spectrum corresponds to low frequencies
# (smooth regions), while edges and fine details appear towards the outside.

# After Gaussian blurring, the spectrum becomes brighter at the center
# and darker at the edges, showing that high-frequency components are reduced.
# This confirms that Gaussian blur works as a low-pass filter.

# In unsharp masking, the outer regions of the spectrum become brighter,
# which indicates enhancement of high-frequency components (edges).

# Increasing the Gaussian sigma or kernel size concentrates energy more
# towards the center of the spectrum, strengthening the low-pass effect.

# Increasing alpha in unsharp masking boosts the high-frequency components,
# leading to stronger sharpening.

````

````python
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

img_str = """

"""

img_data = base64.b64decode(img_str)
img = Image.open(io.BytesIO(img_data))
img_part5 = np.array(img)

plt.imshow(img_part5)
plt.axis("off")
plt.title("Our Image")
plt.show()

import cv2

# RGB -> HSV
hsv = cv2.cvtColor(img_part5, cv2.COLOR_RGB2HSV)

# Yellow colour range in HSV
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([35, 255, 255])

mask = cv2.inRange(hsv, lower_yellow, upper_yellow)

# Change yellow leaf to red
hsv_red = hsv.copy()
hsv_red[mask > 0] = [0, 255, 255]   # Red colour in HSV

# HSV -> RGB
final_img = cv2.cvtColor(hsv_red, cv2.COLOR_HSV2RGB)

plt.imshow(final_img)
plt.axis("off")
plt.title("Yellow Leaf Changed to Red")
plt.show()
````
