````python
import numpy as np

def convolve(img, kernel):

    # Flip kernel for convolution
    kernel = np.flipud(np.fliplr(kernel))

    # Determine padding sizes
    kH, kW = kernel.shape
    pad_h, pad_w = kH // 2, kW // 2

    # Pad the image using reflect mode
    if img.ndim == 2:  # Grayscale
        padded_img = np.pad(img, ((pad_h, pad_h), (pad_w, pad_w)), mode='reflect')
        H, W = img.shape
        out = np.zeros_like(img, dtype=np.float32)
        # Convolve
        for i in range(H):
            for j in range(W):
                out[i, j] = np.sum(padded_img[i:i+kH, j:j+kW] * kernel)

    elif img.ndim == 3:  # RGB
        H, W, C = img.shape
        out = np.zeros_like(img, dtype=np.float32)
        for c in range(C):
            padded_img = np.pad(img[:, :, c], ((pad_h, pad_h), (pad_w, pad_w)), mode='reflect')
            for i in range(H):
                for j in range(W):
                    out[i, j, c] = np.sum(padded_img[i:i+kH, j:j+kW] * kernel)


    out = np.clip(out, 0, 255)
    return out.astype(np.uint8)
````

````python
def average_blur_gray(img, ksize):

    kernel = np.ones((ksize, ksize), dtype=np.float32) / (ksize * ksize)
    return convolve(img, kernel)

def average_blur_rgb(img, ksize):

    kernel = np.ones((ksize, ksize), dtype=np.float32) / (ksize * ksize)
    return convolve(img, kernel)
````

````python
def gaussian_kernel(ksize, sigma):

    ax = np.arange(-(ksize // 2), ksize // 2 + 1)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
    kernel = kernel / np.sum(kernel)
    return kernel
````

````python
def gaussian_blur_gray(img, ksize, sigma):
    kernel = gaussian_kernel(ksize, sigma)
    return convolve(img, kernel)

def gaussian_blur_rgb(img, ksize, sigma):
    kernel = gaussian_kernel(ksize, sigma)
    return convolve(img, kernel)
````

````python

# avg blur
avg_gray = average_blur_gray(gray, 11)
avg_rgb = average_blur_rgb(image, 11)

# Gaussian blur
gauss_gray = gaussian_blur_gray(gray, 11, 1.0)
gauss_rgb = gaussian_blur_rgb(image, 11, 1.0)

# Display results
plt.figure(figsize=(12, 6))
plt.subplot(2,3,1); plt.imshow(gray, cmap='gray'); plt.title("Original Gray"); plt.axis('off')
plt.subplot(2,3,2); plt.imshow(avg_gray, cmap='gray'); plt.title("Average Blur Gray"); plt.axis('off')
plt.subplot(2,3,3); plt.imshow(gauss_gray, cmap='gray'); plt.title("Gaussian Blur Gray"); plt.axis('off')

plt.subplot(2,3,4); plt.imshow(image); plt.title("Original RGB"); plt.axis('off')
plt.subplot(2,3,5); plt.imshow(avg_rgb); plt.title("Average Blur RGB"); plt.axis('off')
plt.subplot(2,3,6); plt.imshow(gauss_rgb); plt.title("Gaussian Blur RGB"); plt.axis('off')
plt.show()
````

````python
def sobel_edges(gray_img):

    # Sobel kernels
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]], dtype=np.float32)

    sobel_y = np.array([[-1, -2, -1],
                        [0, 0, 0],
                        [1, 2, 1]], dtype=np.float32)

    # Convolve with grayscale image
    Gx = convolve(gray_img, sobel_x)
    Gy = convolve(gray_img, sobel_y)

    # Compute gradient magnitude
    gradient_magnitude = np.sqrt(Gx.astype(np.float32)**2 + Gy.astype(np.float32)**2)

    # Normalize to 0-255
    gradient_magnitude = (gradient_magnitude / gradient_magnitude.max()) * 255
    gradient_magnitude = gradient_magnitude.astype(np.uint8)

    return Gx, Gy, gradient_magnitude
````

````python
def threshold_edges(grad_mag, thresh=100):

    edge_map = np.zeros_like(grad_mag)
    edge_map[grad_mag >= thresh] = 255
    return edge_map
````

````python
Gx, Gy, grad_mag = sobel_edges(gray)

# thresholding
edge_map = threshold_edges(grad_mag, thresh=100)


plt.figure(figsize=(12,6))
plt.subplot(2,3,1); plt.imshow(gray, cmap='gray'); plt.title("Original Gray"); plt.axis('off')
plt.subplot(2,3,2); plt.imshow(Gx, cmap='gray'); plt.title("Sobel Gx"); plt.axis('off')
plt.subplot(2,3,3); plt.imshow(Gy, cmap='gray'); plt.title("Sobel Gy"); plt.axis('off')
plt.subplot(2,3,4); plt.imshow(grad_mag, cmap='gray'); plt.title("Gradient Magnitude"); plt.axis('off')
plt.subplot(2,3,5); plt.imshow(edge_map, cmap='gray'); plt.title("Thresholded Edge Map"); plt.axis('off')
plt.show()
````

````python
laplacian_kernel = np.array([[0, 1, 0],
                             [1, -4, 1],
                             [0, 1, 0]], dtype=np.float32)
````

````python
def laplacian_sharpen(img, alpha=1.0):

    lap = convolve(img, laplacian_kernel)
    sharp = img.astype(np.float32) - alpha * lap.astype(np.float32)
    sharp = np.clip(sharp, 0, 255)
    return sharp.astype(np.uint8)
````

````python

# Numpy sharpening
alpha = 1.0
sharp_numpy = laplacian_sharpen(gray, alpha)

# OpenCV sharpening
lap_cv = cv2.Laplacian(gray, cv2.CV_32F)  # Laplacian
sharp_cv = gray.astype(np.float32) - alpha * lap_cv
sharp_cv = np.clip(sharp_cv, 0, 255).astype(np.uint8)

print(sharp_numpy == sharp_cv)
````

````python
def unsharp_mask(img, ksize=5, sigma=1.0, alpha=1.0):

    # Apply Gaussian blur (reuse function from Part A)
    blurred = gaussian_blur_gray(img, ksize, sigma)

    # Compute mask
    mask = img.astype(np.float32) - blurred.astype(np.float32)

    # Add scaled mask
    sharp = img.astype(np.float32) + alpha * mask
    sharp = np.clip(sharp, 0, 255)
    return sharp.astype(np.uint8)
````

````python
sharp1 = unsharp_mask(gray, ksize=11, sigma=1.0, alpha=1.0)   # Sharpened
sharp2 = unsharp_mask(gray, ksize=11, sigma=1.0, alpha=-0.5)  # Blurred effect


plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(gray, cmap='gray'); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(sharp1, cmap='gray'); plt.title("Unsharp Mask alpha=1"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(sharp2, cmap='gray'); plt.title("Unsharp Mask alpha=-0.5"); plt.axis('off')
plt.show()
````

Compute the magnitude spectrum of the original image, gaussian blurred image and the unsharp masking image. Now, play with the alpha value in sharpening and gaussian kernel and notice how the magntude spectrum changes.
Then, in code comments, explain your observations of dark and bright areas in the magnitude spectrum. In nutshell, verify how convolution is actually filtering frequencies depending on the kernel.

````python
def magnitude_spectrum(img):

    # Compute 2D FFT
    f = np.fft.fft2(img)

    # Shift zero frequency component to center
    fshift = np.fft.fftshift(f)

    # Compute magnitude
    magnitude = np.abs(fshift)

    # Log scale for visualization
    magnitude = np.log1p(magnitude)

    # Normalize to 0-255
    magnitude = (magnitude / magnitude.max()) * 255
    return magnitude.astype(np.uint8)
````

````python
# Gaussian blur for LPF
blurred = gaussian_blur_gray(gray, ksize=11, sigma=2.0)

# Unsharp mask for HPF
sharpened = unsharp_mask(gray, ksize=11, sigma=2.0, alpha=1.5)

# Compute magnitude spectra
mag_original = magnitude_spectrum(gray)
mag_blurred = magnitude_spectrum(blurred)
mag_sharpened = magnitude_spectrum(sharpened)
````

````python

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(mag_original, cmap='gray')
plt.title("Original Magnitude Spectrum")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(mag_blurred, cmap='gray')
plt.title("Gaussian Blur (LPF)")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(mag_sharpened, cmap='gray')
plt.title("Unsharp Mask (HPF)")
plt.axis('off')

plt.show()
````

````python
# Observations:
# 1. Original Image:
#    - The magnitude spectrum shows a bright spot at the center (low frequencies)
#      and darker regions towards edges (high frequencies).
#    - Most of the image energy is in low frequencies (smooth variations).

# 2. Gaussian Blur (Low-Pass Filter):
#    - The central area becomes brighter relative to edges.
#    - High frequency components (edges, details) are suppressed,
#      resulting in darker regions away from center.
#    - Increasing sigma makes the blur stronger, further reducing high frequencies.

# 3. Unsharp Masking (High-Pass Filter):
#    - Subtracting Gaussian blur amplifies high frequencies (edges, details).
#    - The outer regions in the magnitude spectrum (high frequencies) become brighter.
#    - Increasing alpha in unsharp masking increases the intensity of high frequencies.
#    - Central low frequencies are less dominant compared to LPF.
````
