````python
# Task 1
def convolve(img, kernel):
    # Get kernel dimensions and calculate padding
    k_h, k_w = kernel.shape
    pad_h = k_h // 2
    pad_w = k_w // 2

    padded_img = np.pad(img,((pad_h, pad_h), (pad_w, pad_w)),mode='reflect')

    conv_img = np.zeros_like(img, dtype=float)

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            patch = padded_img[i:i+k_h, j:j+k_w]
            conv_img[i, j] = np.sum(patch * kernel)

    return conv_img
````

````python
# Task 2
def average_blur_gray(img, ksize):
  kernel = np.ones((ksize, ksize)) / (ksize * ksize)  
  return convolve(img, kernel)  # Using the Function from task 1
````

````python
img_gray = cv2.cvtColor(my_image, cv2.COLOR_RGB2GRAY)
````

````python
# Visualize the results
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(average_blur_gray(img_gray,20), cmap='gray')
plt.title("Average Blurred Grayscale Image")
plt.axis("off")

plt.show()
````

````python
# RGB Blurring
def average_blur_RGB(img,ksize):
    red = img[:,:,0]
    green = img [:,:,1]
    blue = img[:,:,2]

    avg_red = average_blur_gray(red,ksize)
    avg_green = average_blur_gray(green,ksize)
    avg_blue = average_blur_gray(blue,ksize)

    # Combining image channel 
    final_img = np.zeros_like(img)
    final_img[:,:,0] = avg_red
    final_img[:,:,1]= avg_green
    final_img[:,:,2]= avg_blue

    return final_img
````

````python
# Displaying the RGB blurred image
rgb_blurred = average_blur_RGB(my_image,20)

plt.subplot(1,2,1)
plt.imshow(my_image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(rgb_blurred)
plt.title("Average Blurred Image")
plt.axis("off")

plt.show()
````

````python
# Task 3
def gaussian_kernel(ksize, sigma):
  x = np.linspace(-(ksize//2), ksize//2, ksize)
  y = np.linspace(-(ksize//2), ksize//2, ksize)

  # Forming kernel
  gauss_kernel = np.zeros((ksize,ksize))
  for i in range(ksize):
    for j in range(ksize):
      gauss_kernel[i,j] = (1/(2*np.pi*(sigma**2)))*np.exp(-((x[i]**2)+(y[j]**2))/(2*(sigma**2))) 

  return gauss_kernel/np.sum(gauss_kernel)
````

````python
# Display original and Gaussian blured image
gussian_blured_img = convolve(img_gray, gaussian_kernel(9,5.0))

plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(gussian_blured_img, cmap='gray') 
plt.title("Gaussian Blurred Image")
plt.axis("off")

plt.show()
````

````python
# RGB gaussian blur
def gaussian_blur_RGB(img,ksize,sigma):
    red = img[:,:,0]
    green = img [:,:,1]
    blue = img[:,:,2]

    avg_red = convolve(red,gaussian_kernel(ksize,sigma))
    avg_green = convolve(green,gaussian_kernel(ksize,sigma))
    avg_blue = convolve(blue,gaussian_kernel(ksize,sigma))

    # Combining image channel 
    final_img = np.zeros_like(img)
    final_img[:,:,0] = avg_red
    final_img[:,:,1]= avg_green
    final_img[:,:,2]= avg_blue

    return final_img
````

````python
# Function to compute magnitude spectrum
def mag(img):
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    return np.log(abs(fshift))
````

````python
# Yellow Edging Function
def edging_yellow(img):
    red = img[:,:,0]
    green = img [:,:,1]
    blue = img[:,:,2]

    # Sobel on R and G channels
    Rx = convolve(red, sobel_x)
    Ry = convolve(red, sobel_y)
    Gx = convolve(green, sobel_x)
    Gy = convolve(green, sobel_y)

    # Red-Green Channel
    rg = red - green
    rgx = convolve(rg, sobel_x)
    rgy = convolve(rg, sobel_y)

    # Magnitude Calculation
    r_mag = np.sqrt(Rx**2 + Ry**2)
    g_mag = np.sqrt(Gx**2 + Gy**2)
    rg_mag = np.sqrt(rgx**2 + rgy**2)

    edge = rg_mag * 0.6 + r_mag * 0.3 + g_mag * 0.1

    # Normalization
    edging_norm = (edge / edge.max()) * 255
    edging_norm = edging_norm.astype(np.uint8)

    # Binary Thresholding
    edging_map = np.zeros_like(edging_norm)
    edging_map[edging_norm >= 30] = 255

    return edging_map
````
