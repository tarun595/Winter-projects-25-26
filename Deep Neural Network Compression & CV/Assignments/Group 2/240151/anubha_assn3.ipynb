````python
import numpy as np

def convolve(img, kernel):


    # Get image and kernel dimensions
    H, W = img.shape
    kH, kW = kernel.shape

    # Amount of padding needed
    pad_h = kH // 2
    pad_w = kW // 2

    # Pad the image using reflect mode
    padded_img = np.pad(
        img,
        ((pad_h, pad_h), (pad_w, pad_w)),
        mode='reflect'
    )

    # Initialize output image
    output = np.zeros((H, W), dtype=np.float32)

    # Perform convolution
    for i in range(H):
        for j in range(W):
            # Extract local region
            region = padded_img[i:i+kH, j:j+kW]

            # Element-wise multiplication + sum
            output[i, j] = np.sum(region * kernel)

    return output
````

````python
import numpy as np

def average_blur_gray(img, ksize):
    """
    Apply average (box) blur to a grayscale image using NumPy only.

    Parameters:
        img (2D numpy array): Grayscale image
        ksize (int): Kernel size (must be odd)

    Returns:
        blurred (2D numpy array): Blurred image
    """

    assert ksize % 2 == 1, "ksize must be odd"

    H, W = img.shape
    pad = ksize // 2

    # Create box kernel
    kernel = np.ones((ksize, ksize), dtype=np.float32) / (ksize * ksize)

    # Pad image (reflect padding)
    padded_img = np.pad(
        img,
        ((pad, pad), (pad, pad)),
        mode='reflect'
    )

    # Output image
    blurred = np.zeros((H, W), dtype=np.float32)

    # Convolution
    for i in range(H):
        for j in range(W):
            region = padded_img[i:i+ksize, j:j+ksize]
            blurred[i, j] = np.sum(region * kernel)

    blurred = np.clip(blurred, 0, 255).astype(np.uint8)
    return blurred
````

````python
import numpy as np

def average_blur_rgb(img, ksize):
    """
    Apply average (box) blur to an RGB image using NumPy only.

    Parameters:
        img (H, W, 3 numpy array): RGB image
        ksize (int): Kernel size (must be odd)

    Returns:
        blurred (H, W, 3 numpy array): Blurred RGB image
    """

    assert ksize % 2 == 1, "ksize must be odd"
    assert img.ndim == 3 and img.shape[2] == 3, "Input must be RGB image"

    H, W, C = img.shape
    pad = ksize // 2

    # Box kernel
    kernel = np.ones((ksize, ksize), dtype=np.float32) / (ksize * ksize)

    # Output image
    blurred = np.zeros((H, W, 3), dtype=np.float32)

    # Apply blur to each channel separately
    for c in range(3):
        channel = img[:, :, c]

        padded = np.pad(
            channel,
            ((pad, pad), (pad, pad)),
            mode='reflect'
        )

        for i in range(H):
            for j in range(W):
                region = padded[i:i+ksize, j:j+ksize]
                blurred[i, j, c] = np.sum(region * kernel)

    # Convert to displayable format
    blurred = np.clip(blurred, 0, 255).astype(np.uint8)

    return blurred
````

````python
import numpy as np

def gaussian_kernel(ksize, sigma):

    assert ksize % 2 == 1, "ksize must be odd"
    assert sigma > 0, "sigma must be positive"

    # Coordinate grid centered at 0
    ax = np.arange(-(ksize // 2), ksize // 2 + 1)
    xx, yy = np.meshgrid(ax, ax)

    # 2D Gaussian formula
    kernel = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
    kernel = kernel / (2 * np.pi * sigma**2)

    # Normalize so sum = 1
    kernel /= np.sum(kernel)

    return kernel
````

````python
def convolve_gray(img, kernel):
    H, W = img.shape
    k = kernel.shape[0]
    pad = k // 2

    padded = np.pad(img, ((pad, pad), (pad, pad)), mode='reflect')
    out = np.zeros((H, W), dtype=np.float32)

    for i in range(H):
        for j in range(W):
            region = padded[i:i+k, j:j+k]
            out[i, j] = np.sum(region * kernel)

    return np.clip(out, 0, 255).astype(np.uint8)
````

````python
def convolve_rgb(img, kernel):
    H, W, _ = img.shape
    out = np.zeros_like(img, dtype=np.float32)

    for c in range(3):
        out[:, :, c] = convolve_gray(img[:, :, c], kernel)

    return out.astype(np.uint8)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# ----- SOBEL KERNELS -----
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]], dtype=np.float32)

sobel_y = np.array([[-1, -2, -1],
                    [ 0,  0,  0],
                    [ 1,  2,  1]], dtype=np.float32)

# ----- MANUAL CONVOLUTION -----
def convolve2d(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape
    pad = kh // 2
    padded = np.pad(image, pad, mode='constant')
    output = np.zeros((h, w), dtype=np.float32)

    for i in range(h):
        for j in range(w):
            output[i, j] = np.sum(
                padded[i:i+kh, j:j+kw] * kernel
            )
    return output
gray_img = np.array(img.convert("L"))
# ----- SOBEL GRADIENTS -----
Gx = convolve2d(gray_img, sobel_x)
Gy = convolve2d(gray_img, sobel_y)

# ----- GRADIENT MAGNITUDE -----
gradient_magnitude = np.sqrt(Gx**2 + Gy**2)

# ----- NORMALIZATION (0255) -----
edge_map = (gradient_magnitude / gradient_magnitude.max()) * 255
edge_map = edge_map.astype(np.uint8)

# ----- THRESHOLDING -----
threshold = 30
binary_edges = np.zeros_like(edge_map)
binary_edges[edge_map >= threshold] = 255

# ----- DISPLAY OUTPUT -----
plt.figure(figsize=(6,6))
plt.imshow(binary_edges, cmap='gray')
plt.title("Thresholded Sobel Edge Map")
plt.axis('off')
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Helper: Magnitude Spectrum
def magnitude_spectrum(image):
    """
    FFT shifts low frequencies to center.
    Log scale is used because raw FFT values have huge dynamic range.
    Bright center  -> low frequencies
    Bright corners -> high frequencies
    """
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)
    magnitude = np.log(1 + np.abs(fshift))
    return magnitude

# Gaussian Kernel

def gaussian_kernel(size, sigma):
    ax = np.arange(-size//2 + 1., size//2 + 1.)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
    return kernel / kernel.sum()

# Convolution (manual)

def convolve2d(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape
    pad = kh // 2
    padded = np.pad(image, pad, mode='constant')
    output = np.zeros_like(image, dtype=np.float32)

    for i in range(h):
        for j in range(w):
            output[i, j] = np.sum(
                padded[i:i+kh, j:j+kw] * kernel
            )
    return output

# PARAMETERS TO PLAY WITH

gaussian_size = 15
sigma = 3.0            #
alpha = 1.5

# GAUSSIAN BLUR (LOW-PASS)
g_kernel = gaussian_kernel(gaussian_size, sigma)
blurred = convolve2d(gray_img, g_kernel)

# UNSHARP MASKING (HIGH-BOOST)
blurred = convolve2d(gray_img, g_kernel)
mask = gray_img - blurred
sharpened = gray_img + alpha * mask


# MAGNITUDE SPECTRA
mag_original = magnitude_spectrum(gray_img)
mag_blur = magnitude_spectrum(blurred)
mag_sharp = magnitude_spectrum(sharpened)

# DISPLAY

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(gray_img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title("Gaussian Blurred")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(sharpened, cmap='gray')
plt.title("Unsharp Masking")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(mag_original, cmap='gray')
plt.title("Magnitude Spectrum (Original)")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(mag_blur, cmap='gray')
plt.title("Magnitude Spectrum (Blurred)")
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(mag_sharp, cmap='gray')
plt.title("Magnitude Spectrum (Sharpened)")
plt.axis('off')

plt.tight_layout()
plt.show()
````

````python
 import numpy as np
import matplotlib.pyplot as plt

# --- Split RGB channels ---
R = img_part5[:, :, 0].astype(np.float32)
G = img_part5[:, :, 1].astype(np.float32)
B = img_part5[:, :, 2].astype(np.float32)

# --- Key idea: exploit R - G transition ---
RG_diff = R - G
RG_diff = (RG_diff - RG_diff.min()) / (RG_diff.max() - RG_diff.min())

# --- Custom edge kernels ---
kernel_h = np.array([
    [-1, -1, -1],
    [ 0,  0,  0],
    [ 1,  1,  1]
], dtype=np.float32)

kernel_v = kernel_h.T


# --- Manual convolution ---
def convolve2d(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape
    pad = kh // 2
    padded = np.pad(image, pad, mode='edge')
    out = np.zeros_like(image)

    for i in range(h):
        for j in range(w):
            out[i, j] = np.sum(
                padded[i:i+kh, j:j+kw] * kernel
            )
    return out


# --- Apply edge detection on RG difference ---
edge_h = convolve2d(RG_diff, kernel_h)
edge_v = convolve2d(RG_diff, kernel_v)

edge_map = np.sqrt(edge_h**2 + edge_v**2)

# --- Normalize ---
edge_map = (edge_map / edge_map.max()) * 255
edge_map = edge_map.astype(np.uint8)

# --- Threshold for clean outline ---
binary_edges = np.zeros_like(edge_map)
binary_edges[edge_map > 40] = 255


# --- Display ---
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_part5)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(RG_diff, cmap='gray')
plt.title("R  G Feature Map")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(binary_edges, cmap='gray')
plt.title("Yellow Transition Edges")
plt.axis("off")

plt.show()
````
