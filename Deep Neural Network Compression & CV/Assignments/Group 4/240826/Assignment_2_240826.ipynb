````python
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = """ """
  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
````

````python
img_rgb=np.array(img)
img_gray= (0.299 * img_rgb[:, :, 0] + 0.587 * img_rgb[:, :, 1] + 0.114 * img_rgb[:, :, 2])
````

````python
def gray_histogram(img_gray):
  hist = np.zeros(256, dtype=int)
  m,n=img_gray.shape
  for i in range(m):
    for j in range(n):
        hist[int(img_gray[i,j])]+=1
  return hist
````

````python
import matplotlib.pyplot as plt
hist = gray_histogram(img_gray)

plt.plot(hist)
plt.title("Grayscale Image Histogram")
plt.xlabel("Pixel value")
plt.ylabel("Frequency")
plt.show()
````

````python
def rgb_histogram(img_rgb):
  hist = np.zeros((3, 256), dtype=int)
  m,n,c=img_rgb.shape
  for i in range(m):
    for j in range(n):
        hist[0,img_rgb[i,j,0]]+=1
        hist[1,img_rgb[i,j,1]]+=1
        hist[2,img_rgb[i,j,2]]+=1
  return hist
````

````python
hist=rgb_histogram(img_rgb)
plt.plot(hist[0], color='r')
plt.plot(hist[1] ,color='g')
plt.plot(hist[2], color='b')
plt.title("RGB Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
````

````python
def rgb_to_hsv(img_rgb):
  img_hsv=np.zeros(img_rgb.shape)
  m,n,c=img_rgb.shape
  for i in range(m):
    for j in range(n):
      R=img_rgb[i,j,0]
      G=img_rgb[i,j,1]
      B=img_rgb[i,j,2]
      r=R/255
      g=G/255
      b=B/255
      Cmax=max(r,g,b)
      Cmin=min(r,g,b)
      delta=Cmax-Cmin
      if(delta==0):
        H=0
      elif(Cmax==r):
        H=60*(((g-b)/delta)%6)
      elif(Cmax==g):
        H=60*(((b-r)/delta)+2)
      elif(Cmax==b):
        H=60*(((r-g)/delta)+4)
      if(Cmax==0):
        S=0
      else:
        S=delta/Cmax
      V=Cmax
      img_hsv[i,j,0]=H/2
      img_hsv[i,j,1]=S*255
      img_hsv[i,j,2]=V*255
  return img_hsv.astype(np.uint8)
````

````python
def change_saturation(img_rgb, factor):
    hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV).astype(np.float32)
    hsv[...,1] *= factor
    hsv[...,1] = np.clip(hsv[...,1], 0, 255)
    return cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2RGB)
````

````python
def clamp(img):
    return np.clip(img, 0, 255).astype(np.uint8)
````

````python
def warm_pizza(img_rgb):
    img = img_rgb.astype(np.float32)

    # 1) Slight brightness + contrast
    img = img * 1.08 + 10    # small contrast + small brightness
    img = clamp(img)

    # 2) Slightly increase saturation
    img = change_saturation(img, 1.15)

    # 3) Warm tint: increase red a little, reduce blue a little
    img = img.astype(np.float32)
    img[...,0] *= 1.05   # red channel
    img[...,2] *= 0.95   # blue channel
    img = clamp(img)

    return img
````

````python
def crispy_pizza(img_rgb):
    img = img_rgb.astype(np.float32)

    # 1) VERY small contrast increase
    img = img * 1.05 + 5   # far more gentle
    img = clamp(img)

    # 2) Small saturation increase
    img = change_saturation(img, 1.08)

    # 3) sharpen:
    #    sharpen = img + 0.1 * (img - blur)
    blur = cv2.GaussianBlur(img, (7,7), 0)
    sharpen = img + 0.1 * (img - blur)   # â† low sharpen factor
    img = clamp(sharpen)

    return img
````

````python
out1 = warm_pizza(img_rgb)
out2 = crispy_pizza(img_rgb)
````

````python
plt.figure(figsize=(14,6))
plt.subplot(1,3,1); plt.imshow(img_rgb); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(out1); plt.title("WarmPizza"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(out2); plt.title("CrispyPizza"); plt.axis('off')
plt.show()
````

````python
def white_patch_balance(img_rgb):
  R_max = np.max(img_rgb[..., 0])
  G_max = np.max(img_rgb[..., 1])
  B_max = np.max(img_rgb[..., 2])
   # avoid division by zero
  if R_max==0:R_max=1
  if G_max==0:G_max=1
  if B_max==0:B_max=1

  R_scale = 255.0 / (R_max)
  G_scale = 255.0 / (G_max)
  B_scale = 255.0 / (B_max)

  out = np.zeros_like(img)
  out[..., 0] = img_rgb[..., 0] * R_scale
  out[..., 1] = img_rgb[..., 1] * G_scale
  out[..., 2] = img_rgb[..., 2] * B_scale
  out = np.clip(out, 0, 255).astype(np.uint8)

  return out

````
