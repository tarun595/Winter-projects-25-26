````python
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = """"""

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
````

````python
rgb=np.array(img)
````

````python
gray_np = (0.299 * rgb[:, :, 0] + 0.587 * rgb[:, :, 1] + 0.114 * rgb[:, :, 2])
gray_np = gray_np.astype(np.uint8)
plt.imshow(gray_np, cmap="gray")
plt.title("Grayscale (NumPy Weighted Sum)")
plt.axis("off")
plt.show()
````

````python
def gray_histogram(gray_np):
  hist=np.zeros(256)
  h,w=gray_np.shape
  for i in range(h):
    for j in range(w):
        hist[gray_np[i,j]]+=1
  return hist
````

````python
plt.plot(gray_histogram(gray_np))
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("Grayscale Histogram")
plt.show()
````

````python
def rgb_histogram(rgb):
  hist=np.zeros((3,256))
  h,w,c=rgb.shape
  for i in range(h):
    for j in range(w):
      for k in range(c):
        hist[k,rgb[i,j,k]]+=1
  return hist
````

````python
hist=rgb_histogram(rgb)
````

````python
plt.plot(hist[0], color='r', label='Red')
plt.plot(hist[1], color='g', label='Green')
plt.plot(hist[2], color='b', label='Blue')
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("RGB Histogram")
plt.show()
````

````python
def rgb_to_hsv(rgb):
  hsv=np.zeros(rgb.shape)
  h,w,c=rgb.shape
  for i in range(h):
    for j in range(w):
      r=rgb[i,j,0]/255
      g=rgb[i,j,1]/255
      b=rgb[i,j,2]/255
      max_val=max(r,g,b)
      min_val=min(r,g,b)
      diff=max_val-min_val
      if diff==0:
        h=0
      elif max_val==b:
        h=60*(((r-g)/diff)+4)
      elif max_val==g:
        h=60*(((b-r)/diff)+2)
      elif max_val==r:
        h=60*(((g-b)/diff)%6)
      if max_val==0:
        s=0
      else:
        s=diff/max_val
      v=max_val
      hsv[i,j,0]=h/2
      hsv[i,j,1]=s*255
      hsv[i,j,2]=v*255
  return hsv.astype(np.uint8)
````

````python
img_hsv=rgb_to_hsv(rgb)
plt.imshow(img_hsv)
plt.axis("off")
````

````python
def clip(img):
    return np.clip(img, 0, 255).astype(np.uint8)
def IceCream_Pop(rgb, contrast=1.15, brightness=15, sat_scale=1.35):
    out = rgb.astype(np.float32)

    # (1) increase contrast and brightness
    out = out * contrast + brightness
    out = clip(out)

    # (2) boost saturation
    hsv = cv2.cvtColor(out, cv2.COLOR_RGB2HSV).astype(np.float32)
    hsv[:, :, 1] = np.clip(hsv[:, :, 1] * sat_scale, 0, 255)
    out = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2RGB)

    return out
````

````python
def Minty_Freeze(rgb, gamma=0.9, tint_strength=0.12):
    out = rgb.astype(np.float32) / 255.0

    # (1) gamma correction to soften image
    out = (out ** (1.0 / gamma)) * 255.0

    # (2) add cool blue tint
    cool = np.array([120, 200, 255], dtype=np.float32)
    out = out * (1 - tint_strength) + cool * tint_strength

    return clip(out)
````

````python
f1 = IceCream_Pop(rgb)
f2 = Minty_Freeze(rgb)
````

#### White Balance Function
Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image. Use Numpy only to keep track of the brightest pixel and then scale the entire image with the factor you find. Display the original and output image.

````python
# Your PART D codes follow here. You can create more cells under this. Write clean and commented code.
````

````python
def white_patch_balance(img_rgb):
    # convert to float for safe computation
    img = img_rgb.astype(np.float32)

    # find maximum value in each channel
    max_r = np.max(img[:, :, 0])
    max_g = np.max(img[:, :, 1])
    max_b = np.max(img[:, :, 2])

    # avoid division by zero
    if max_r == 0: max_r = 1
    if max_g == 0: max_g = 1
    if max_b == 0: max_b = 1

    # scale each channel so brightest becomes 255
    img[:, :, 0] = img[:, :, 0] * (255 / max_r)
    img[:, :, 1] = img[:, :, 1] * (255 / max_g)
    img[:, :, 2] = img[:, :, 2] * (255 / max_b)

    # clip values to valid range
    img = np.clip(img, 0, 255)

    return img.astype(np.uint8)
````

````python
balanced = white_patch_balance(rgb)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("White Patch Balanced")
plt.imshow(balanced)
plt.axis("off")

plt.show()
````
